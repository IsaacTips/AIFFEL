## 1. 인공지능 개발자가 되기 위한 위대한 첫 걸음

앞으로 많이 접하게 될 개발 환경에 대해 공부해볼 것이다. 가장 먼저, 다룰 개발 환경이 어떤 것인지 알아보자. 공부하거나 과제 실습을 진행할 개발 환경은 다음과 같다.

* Ubuntu 18.04
* Python 3.7
* TensorFlow 2.2.0


### (1) Ubuntu 18.04
---
첫 번째, 우분투(Ubuntu)는 리눅스 기반으로 만들어진 운영 체제를 뜻한다. 리눅스, 운영 체제라.. 굉장히 컴퓨터 공학스러운 단어들이다. 운영체제에 대해 재미있게 설명하는 글을 읽으며 가볍게 알아보자.

[운영체제(Operating System : OS)란 ? 운영체제 종류와 목적, OS 정복](https://jhnyang.tistory.com/16)

간단히 말하자면 운영 체제는 컴퓨터가 처리해야 하는 복잡하면서도 공통적인 기능들을 알아서 처리해주어 사용자가 신경쓰지 않아도 되도록 만들어주는 프로그램이라고 할 수 있다. 대표적인 예로, 컴퓨터를 켤 때마다 원래 저장되어 있던 파일이나 폴더 등을 하드디스크에서 꺼내서 쓸 수 있도록 준비하지 않아도 된다. 운영 체제 내의 부팅을 관리하는 프로그램이 알아서 바탕화면에서 바로 사용할 수 있도록 부팅을 할 때마다 준비해주기 때문이다.

가장 많이 쓰이는 운영 체제로는 마이크로소프트에서 만든 Windows, 애플에서 만든 MacOS, 그리고 리눅스 토발즈가 만든 Linux 등이 있다. 우분투는 바로 이 Linux를 기반으로 만들어진 운영체제이다. Linux를 기반으로 또다른 운영체제가 만들어질 수 있었던 이유는 바로 Linux가 자유롭게 수정 및 재배포가 될 수 있는 오픈소스기 때문이다. 이와 달리 기업에서 만들어진 Windows와 MacOS는 자유롭게 수정 및 재배포 할 수 없다.

사실 오픈소스라는 특성 때문에 Linux로부터 파생된 운영체제는 우분투 외에도 Fedora, RedHat 등 무척 다양하다. 이 중 우분투는 개인용, 또는 데스크탑 환경에 최적화 된 버전이다. 그렇다면 우분투 뒤에 있는 18.04 라는 숫자는 무엇을 의미할까?

우분투에 간단히 소개하는 다음 글을 하나 더 읽어보면서 답을 찾아보겠다.

[Ubuntu(우분투)란? 무엇인가](https://dongdong-2.tistory.com/17)

보통 많은 프로그램에서 뒤에 붙는 숫자는 버전을 나타낸다. 버전을 나타내는 숫자가 꼭 배포된 날짜와 연관되어 있는 것은 아니다. 하지만 우분투는 특별히 날짜와 연관시켜 버전 번호를 매기며, 18.04는 2018년 4월에 나온 버전이라는 뜻을 갖고 있다.

### (2) Python 3.7
---
두 번째, Python은 프로그래밍 언어를 뜻한다. 개발 또는 프로그래밍을 처음 접할 때 가장 먼저 선택하게 되는 것은 주로 프로그래밍 언어일 것이다. 어떤 언어로 개발을 시작하냐에 따라 나중에 깊게 들어갈 개발 분야가 크게 달라질 수도 있다.

프로그래밍 언어에 관한 흥미로운 영상 하나를 보도록 하겠다. 지금 당장 모든 언어들에 대해 자세히 알기보다는, 언어들의 생태계를 이해해보자.

[![노마드 코더](https://img.youtube.com/vi/j5ogDOuqXqg/0.jpg)](https://www.youtube.com/watch?v=j5ogDOuqXqg&feature=emb_logo&ab_channel=%EB%85%B8%EB%A7%88%EB%93%9C%EC%BD%94%EB%8D%94NomadCoders)

보다 __범용적인 목적으로 만들어진 언어__가 있는 반면, __특정한 문제를 풀기 위한 언어__도 존재한다. 파이썬은 영상에서 설명했듯 범용적인 목적으로 만들어진 언어이다. 비행기 진단이나, 병렬 처리 등을 위해 최적화된 것보다는 사용자가 원하는대로 아주 다양한 문제들을 풀 수 있도록 설계된, 자유도가 높은 언어인 것이다. 특히 AI 분야에서 파이썬은 매우 활발히 활용되고 있다. 머신러닝/딥러닝과 관련된 다양한 라이브러리들이 파이썬으로 개발되었고, 지금도 활발하게 성장하고 있기 때문이다.

Python 뒤에 붙은 3.7은 버전을 나타낸다. 소프트웨어 버전에서 소수점 이하 숫자가 바뀌는 것은 작은 변화를, 앞자리가 바뀌는 것은 큰 변화를 의미한다. Python 2와 3은 꽤나 많은 차이를 가지는데, 2020년 현재 2.x 버전은 이제 거의 없어지고 있고, 대부분의 파이썬 코드가 3.x 버전으로 쓰이고 있다.

### (3) TensorFlow 2.2.0
---
마지막은 텐서플로우(TensorFlow)이다. AI, 인공지능에 관심 있는 사람이라면 한 번쯤 들어봤을 것이다. 바로 머신러닝/딥러닝에 특화된 라이브러리이다.

여기서 잠깐, 라이브러리(library) 란 무엇일까? 라이브러리란, __특정 기능을 위한 여러 함수 또는 클래스를 담고 있는 보따리__라고 설명할 수 있다. 즉, 특정 목적을 위해 필요한 비슷한 함수, 또는 클래스들을 편리하게 한 번에 꺼내어 쓸 수 있도록 한 곳에 모아둔 개념이다. 라이브러리와 비슷한 개념의 단어로는 패키지(package), 모듈(module) 등이 있는데, 모듈은 조금 더 작은 개념으로 쓰이기도 한다.

따라서, 다시 말하자면 텐서플로우는 머신러닝/딥러닝을 위해서 공통적으로 많이 쓰이는 함수 또는 클래스들을 한 곳에 모아놓은 보따리라고 볼 수 있겠다.

사실 머신러닝/딥러닝에 특화된 라이브러리는 텐서플로우 말고도 파이토치(PyTorch), 케라스(Keras) 등 여러가지가 있다. 각 라이브러리는 장단점이 있고, 분야에 따라 조금 더 선호되는 정도가 다르다. 인공지능을 공부하는 사람들 사이에서 라이브러리에 대한 선호가 다를 수 있으니, 다른 사람들과 라이브러리에 대한 생각을 나누어 이야기 해보는 것도 좋다. 보통 많은 사람들이 특별히 좋아하는 라이브러리가 있어서, 각 라이브러리에 대한 장점을 마구 알려줄 수도 있기 떄문이다.

텐서플로우의 버전은 2.2.0 을 사용할 예정이다. 텐서플로우는 0.x 버전으로 시작해 2017년 1.0 알파 버전을 발표한 이후, 오랫동안 1.x 버전을 유지하다가 출시 2년이 넘은 지난 2019년 9월에 2.0 정식 버전을 내놓았다. 1.x 버전에서 2.x 버전으로 넘어오면서 몇 가지 큰 변화가 있었다.

다음의 공식 텐서플로우 문서를 한번 살펴보자!

[Effective TensorFlow 2 | TensorFlow Core](https://www.tensorflow.org/guide/effective_tf2)

가장 중요한 흐름은 훨씬 직관적이면서도 편리해졌다는 점이다. 여러 가지로 분산되어 있던 API를 통합해 일관성을 키우고, 복잡한 코드가 필요했던 부분을 제거하여 깔끔하게 사용할 수 있게 하는 등 사용자를 위한 변화가 돋보인다.

핵심만 정리하자면 지금까지 많이 접했을 **Windows**나 **MacOS**가 아닌, 우분투라는 새로운 운영체제를 사용한다는 것이다. 우분투 운영체제로 구성된 컴퓨터는 이전에 익숙하던 형태와는 다르기 때문에 간단한 것을 다루는 것부터 어색할 수 있다. 하지만 우분투는 사용자의 편의에 초점을 맞춰 개발된 운영체제로 편리한 사용이 특징인만큼, 익숙해지는 데에 오래 걸리지 않을 것이다.

## 2. 까만 화면에 흰 글씨, 해커만 쓰는 거 아니냐고? 나도 쓸 수 있다!
프로그래밍 또는 해커! 하면 뭐가 떠오르는가? 보편적으로 까만 화면에 흰 글씨가 막 빠르게 적혀내려가는 그런 이미지를 많이 떠올린다. 뭔가 대단한 것 같지만, 사실 생각보다 간단하다. 이번 스텝에서는 이러한 것들을 처음 접해보고, 여러 가지로 다루는 법을 공부해볼 것이다. 먼저 그 까만 화면이 무엇인지 알아보겠다.

<img src="./image/terminal1.png" alt="Terminal" width="50%"/>

저 화면의 공식 이름은 바로 __터미널(Terminal)__이다.

기본 값으로 배경화면이 까만색으로 설정되어 있기는 하지만, 사실 원하는대로 바꿀 수 있는 커스터마이징 기능을 지원하는 터미널도 많다. 우리가 사용하는 Ubuntu 운영체제에서는 간단한 단축키로 터미널으르 실행시킬 수 있다.

바로 `Ctrl`+`Alt`+`T` 이다! 바로 한 번 실행해보자! 다음과 같은 화면이 나왔을 것이다. 직접 터미널 프로그램을 찾아서 실행시킬수도 있지만, 앞으로 빠른 실행을 위해 `Ctrl`+`Alt`+`T`를 기억하자.

주피터 노트북 실행을 위해 켜져 있는 터미널이 하나 있다면, '새로운 탭' 또는 '새로운 창'을 선택해 터미널을 한 개 더 열자.

<img src="./image/terminal2.png" alt="Terminal" width="50%"/>

위와 같은 터미널 환경을 __CLI(Command Line Interface)__ 라고 한다. 직역하면 "명령 줄 장치"라고 할 수 있겠다.

이 화면이 대체 어떤 것을 위한 프로그램인지 쉽게 감이 오지 않을 수 있지만, 사실 CLI는 간단히 말하자면 Windows의 탐색기, 또는 Mac의 파인더(Finder)와 비슷한 역할을 한다. 즉, 폴더나 파일을 관리하거나 실행하고, 또는 조금 더 나아가 여러 명령을 통해 복잡한 작업들을 수행할 수 있도록 한다는 것이다.

다만 Windows의 탐색기나 Mac의 파인더는 눈으로 폴더를 볼 수 있으며 마우스 클릭을 통해 원하는 동작들을 수행하는 반면, CLI 환경에서는 키보드로 작성하는 명령어를 통해 원하는 동작을 수행한다. 그렇기 때문에 이름도 "Command-Line Interface"라고 하게 된 것이다. 반대로, 전자처럼 사용자가 눈으로 확인하며 작업할 수 있는 환경은 GUI(Graphical User Interface)라고 한다. 그래픽으로 표현된다는 뜻을 담고 있다.

먼저 앞으로 주로 사용할 프로그래밍 언어를 실행시켜보겠다. 사용할 프로그래밍 언어, 파이썬, 실행해보자.

```
$ python3
```

아래와 같은 결과가 나왔다면 성공이다.

```python
Python 3.7.0 (default, Dec 28 2018, 19:18:39)
[Clang 10.0.0 (clang-1000.10.44.4)] on darwin
Type "help", "copyright", "credits" or "license" for more information.
>>>
```

결과로 나온 처음 세 줄은 파이썬이 실행되면서 나온 안내문이다. 앞서 알아봤던 대로 3.7.0 버전을 사용하는걸 확인할 수 있다. 맨 마지막에 나온 `>>>` 기호는 명령을 기다리고 있다는 표시이다. 이제부터는 파이썬 문법에 맞는 명령어를 써야한다!

파이썬 하면 어느 누구나 가장 먼저 한다는 `Hello World`, 한번 해보자.

```python
>>> print("Hello World!")
Hello World
```

잘 출력되는 모습이다. 그렇다면 이번에는 자주 사용하는 딥러닝 프레임워크, 텐서플로우를 한번 불러와 보자.

```python
>>> import tensorflow as tf
```

첫 줄의 `import tensorflow as tf` 는 앞으로 굉장히 많이 쓸 코드이다. tensorflow라는 라이브러리를 tf 라는 이름으로 가져와줘! 라는 뜻이다. 어떤 라이브러리를 가져오더라도 as 를 사용하면 tensorflow처럼 코드로 작성하기 어려운 긴 이름 대신 간단하면서도 짧은 이름으로 사용할 수 있다.

문제 없이 가져와 졌다면, 한번 버전을 출력해보자. `tf.version` 을 출력하면 현재 사용하는 텐서플로우의 버전을 확인할 수 있다.

```python
>>> print(tf.__version__)
2.2.0
```

이렇게 터미널 상에서 파이썬을 바로 실행해서 사용할 수 있다. 명령을 기다리고 있는 쉘을 종료하고 싶다면, `exit()` 또는 `quit()` 을 입력하면 된다.

```python
>>> exit()
```

하지만 여기서 문제가 있다. 터미널 상에서 파이썬을 사용하면 코드가 한 줄 단위로 실행되기 때문에, 더욱 복잡한 코드를 다루기에는 분명한 한계가 있다. 이러한 한계를 극복하기 위해, 세상에는 다양한 __IDE(Integrated Development Environment)__ 이 존재한다. IDE란 통합 개발 환경의 줄임말 로, 다양한 코드 편집 기능이 있어서 코드 작성을 편리하게 해주고, 완성된 코드들을 자동으로 빌드해주며, 문제가 있을 경우 디버깅 또한 편리하게 할 수 있도록 도와주는 프로그램이다.

IDE는 목적에 따라 성격이 조금씩 다른 여러 종류가 있다. 그렇다면 가장 많이 쓰이는 파이썬 IDE는 무엇이 있을까? 다음 글에서 확인해보자.

[12 Best Python IDEs and Code Editors in 2020](https://www.softwaretestinghelp.com/python-ide-code-editors/)

아주 다양한 IDE가 존재한다는 것을 알 수 있다. 이미지도 유심히 보았다면, 그 화면들이 코딩을 하기에 최적화된 화면인 것을 확인하실 수 있다.

가장 많이 쓰이는 IDE는 위와 같지만, 개인의 취향에 따라 각각 선호하는 IDE가 서로 다른 경우도 많다. 무엇보다, 다양한 것들을 경험해보고 나에게 맞는 것을 찾는 것이 좋다.

### (1) 가장 첫번째 명령, 내 이름을 보여줘!

```
$ whoami
```
`whoami` 명령어를 치면 현재 컴퓨터의 유저 이름이 출력된다.

### (2) 내 현재 위치는 어디야? 경로를 출력해줘!
---
```
$ pwd
/home/username
```

현재 터미널에서 위치해있는 경로를 알고 싶을 때에는 `pwd` 명령어를 입력하면 된다. 일반적으로 터미널을 켠 직후 현재 위치를 출력하면, 위와 같이 `/home` 밑에 내가 사용하고 있는 유저 계정의 최상단 디렉토리에 위치해 있을 것이다. 가장 기본이 되는 이 위치는 `Home` 디렉토리라고 부르기도 하고, `~` 라는 약어로 나타내기도 한다. 참고로, __디렉토리(Directory)__ 는 일반적으로 이야기하는 폴더와 같은 단어이다.

`pwd` 는 Print Working Directory, 즉 "현재 디렉토리(폴더)를 출력하라"라는 뜻의 줄임말이다. 꽤나 직관적이다. 이렇게 각 명령어가 어떤 말의 줄임말인지를 아는 것은 더 많은 명령어를 쉽게 외우는 것에 도움이 될 것이다.

### (3) 현재 디렉토리 내에 있는 모든 파일 목록을 보여줘!
---
```
$ ls
Applications
Desktop
Downloads
Music
...
```

`ls` 는 List의 줄임말로, 현재 디렉토리 내의 모든 파일 또는 하위 디렉토리의 목록을 모두 출력한다. GUI 환경에서는 특정 폴더를 더블클릭해서 폴더에 진입하게 되면 그 폴더 안에 어떤 것들이 있는지 그래픽으로 바로 볼 수 있지만, CLI 환경에서는 그렇지 않다. 그래서 현재 폴더에 어떤 파일 또는 다른 폴더가 있는지 보기 위해서는 ls 명령어를 통해 확인해야 한다.

각 파일 또는 폴더에 대한 좀 더 자세한 목록을 보고싶다면 약간의 옵션을 추가해주면 된다.

```
$ ls -al
drwxr-xr-x+  79 minn  staff    2528  2 27 17:17 .
drwxr-xr-x    5 root  admin     160 10 24 09:37 ..
drwxr-xr-x    3 minn  staff      96 11 10 19:37 .vim
drwx------@   5 minn  staff     160  2 20 17:19 Applications
drwx------@  10 minn  staff     320  2 17 22:32 Desktop
drwx------@  24 minn  staff     768  2 27 04:09 Downloads
drwxr-xr-x    7 minn  staff     224  2 17 22:14 Music
...
```

위와 같이 조금 더 복잡하고 이전에는 나오지 않았던 목록들이 출력되었다. 기본 명령어 뒤에 `-` (dash) 하나를 입력 후 추가 명령어를 입력하면 세부 옵션을 설정할 수 있다. `-` 뒤의 `a` 는 숨김 파일까지 모두 출력하라는 옵션이고, `l` 은 자세히(long format) 정보를 출력하라는 의미를 가지고 있다. 터미널에서 숨김파일은 파일/폴더명 앞에 `.` 이 붙게 되므로, 이전에는 나오지 않았던 여러 목록들까지 출력이 된 것을 확인할 수 있다. 각 줄에 길고 복잡하게 출력된 정보는 어떤 의미를 가질까? 다음 글에서 확인해보자.

[리눅스 ls -l 명령어 - 파일 정보 확인과 의미](https://www.leafcats.com/137)

### 4) 원하는 디렉토리로 이동해보자! 더블클릭!
```
---
$ cd Downloads
$ pwd

/home/username/Downloads
```

GUI 환경에서 원하는 디렉토리로 이동하고 싶을 때는 눈에 보이는 폴더를 더블클릭해서 들어가야 했다. 하지만 앞서 언급했듯 CLI 환경에서는 더블클릭을 할 수 없기 때문에 폴더를 이동하는 것 또한 명령어로 진행해야 한다. 그걸 하게 해주는 명령어가 바로 `cd` 이다. `cd` 는 Change Directory, 즉 디렉토리 변경의 줄임말이다. `cd` 명령어 뒤에 이동하기를 원하는 디렉토리 이름을 입력하면 그 디렉토리로 이동한다.

```
$ cd ..
$ pwd

/home/username
```

`cd` 뒤에 원하는 디렉토리의 이름 대신, `..` 을 입력하면 상위 폴더로 나가게 된다.

```
$ cd ~
$ pwd

/home/username
```

`cd ~` 라는 명령어는 `~` 위치로 이동시키라는 명령어로써, 현재 어떤 폴더 위치에 있었든지 항상 `/home/username` 위치로 이동할 수 있다. 여러 디렉토리 안으로 깊게 들어갔는데 한 번에 Home 디렉토리로 돌아오고 싶을 때 유용한 명령어이다.

### (5) 새 폴더 만들기, 또는 폴더 삭제하기
---
폴더를 이동하는 것뿐만 아니라, 원하는 폴더를 만들거나 삭제할 수도 있어야한다.

```
$ mkdir new_folder
$ ls

Applications
Desktop
Downloads
Music
new_folder
...
```

`mkdir` 은 직관적으로 봐도 Make Directory로 보인다. `new_folder`라는 이름의 폴더를 생성한 후 `ls` 명령어로 목록을 다시 출력하면 `new_folder` 라는 디렉토리가 새로 생긴 것을 확인할 수 있다.

그렇다면 폴더 삭제는 어떻게 할까?

```
$ rm -r new_folder
$ ls

Applications
Desktop
Downloads
Music
...
```

`rm -r` 명령어 뒤에 삭제하길 원하는 디렉토리 명을 입력하면 삭제가 된다. `rm`은 Remove의 줄임말이다. `-r` 옵션은 디렉토리를 삭제할 때 필요한데, 하위 디렉토리와 파일까지 모두 삭제하라는 것을 의미한다. 디렉토리가 아닌 일반 파일을 삭제할 때에는 `rm` 명령어만 입력하면 된다.

참고로 `-r` 옵션은 Recursive라는 의미로, 디렉토리 내부의 모든 파일 및 폴더에 대해 재귀적(반복적)으로 명령을 수행하라는 의미를 가진다. 그렇기 때문에 디렉토리에게는 `-r` 옵션이 필요하고, 개별 파일에게는 `-r` 옵션이 필요 없는 것이다

### (6) 파일을 원하는 곳으로 옮기거나, 복사하기
---
CLI 환경에서는 어떻게 파일이나 디렉토리를 옮길 수 있을까?

```
$ cd ~
$ mkdir aiffel
$ mkdir new_folder
$ mv new_folder isaac
```

무슨 일이 벌어진걸까? `new_folder` 디렉토리를 새로 만들어준 후, `mv` 명령어를 이용해서 `new_folder` 디렉토리를 `isaac` 안으로 옮겼다.

`mv` 명령어는 영어 단어 Move를 줄여서 만든 것임을 어렵지 않게 추측할 수 있다. 파일이나 디렉토리를 옮기고 싶을 때에는 `mv` 명령어 뒤에 이동하고 싶은 파일, 이동할 목적지 디렉토리를 순서대로 입력해주면 된다. 잘 옮겨졌는지 `isaac`로 가서 확인해보자.

```
$ cd isaac
$ ls

new_folder
```

`ls` 를 입력해서 `new_folder` 가 목록에 나온다면 성공한 것이다. 같은 원리로 파일 또는 디렉토리를 복사하는 것도 매우 간단하다.

```
$ cp -r new_folder ..
$ cd ..
ls

Applications
Desktop
Downloads
Music
new_folder
...
```

첫번째 명령어에서 `new_folder` 를 `..` 의 위치, 즉 상위 폴더에 복사했다. 물론 `cp` 는 Copy의 줄임말이다.

다만, 여기서 한 가지 주의할 점은 `cp` 명령어는 `rm`(삭제) 명령어와 같이 디렉토리를 복사할 때 `-r` 옵션을 추가해주어야 복사하려는 디렉토리의 하위디렉토리까지 함께 복사한다는 점이다. 개별 파일을 복사하고 싶을 때에는 `-r` 이 필요없다.


## 4. Terminal, 또 무엇을 할 수 있을까?
이번 단계에서는 터미널에서 할 수 있는 좀 더 심화된 동작을 알아보겠다.

터미널에서 할 수 있는 중요한 한 가지는 패키지 설치 및 관리이다.

__패키지(Package)__ 란, 간단히 말해서 특정 기능을 하는 작은 프로그램 단위이다. 

개발을 하다보면 다양한 패키지들이 필요하다. 패키지는 우리가 다운 받아 사용할 수 있는 다양한 툴을 제공하기 때문에, 효율적인 개발 작업을 위해서는 많은 패키지들을 설치하거나 삭제하는 등 자유자재로 관리할 수 있어야 한다.

우분투에서 패키지를 관리하기 위해 주로 쓰이는 명령어는 `apt-get` 이다. 다음 글을 통해 명령어를 확인해보자.

[Ubuntu 18.04 LTS apt-get 명령어 사용법](https://snrnsnrnd.tistory.com/12)

직접 해보도록 하자. 먼저 지금까지 설치된 패키지의 리스트를 한 번 확인해보자.

```
$ sudo apt list --installed

Listing...
acl/xenial,now 2.2.52-3 amd64 [installed]
adduser/xenial,xenial,now 3.113+nmu3ubuntu4 all [installed]
...
```

`sudo apt list --installed` 명령어를 입력하면 설치된 패키지의 목록이 차례로 출력된다.

여기서 추가 팁 하나!

```
$ sudo apt list --installed | grep acl

acl/bionic,now 2.2.52-3build1 amd64 [installed]
...
```

`sudo apt list --installed` 명령어 뒤에 `| grep` (패키지명) 을 입력해주면 검색하는 단어를 포함하는 패키지만 출력된다. 보통 아주 많은 패키지가 설치되어 있기 때문에 원하는 패키지의 설치 여부를 확인하려면 눈이 빠지게 리스트를 조사해야 한다. 이럴 때 grep 옵션을 사용하면 편리하게 원하는 패키지만 검색해 볼 수 있다.

컴퓨터에 설치된 패키지 인덱스 정보를 업데이트 하려면 다음 명령어를 실행한다. 아래 명령어는 주기적으로 실행시키는 습관을 가지는 것이 좋다.

```
$ sudo apt-get update
```

다음 명령은 모든 패키지에 대해, 새롭게 업데이트 된 버전이 있다면 전부 업그레이드를 하는 명령어이다. 하지만 패키지를 최신 버전으로 업그레이드 하는 것은 언제나 기존에 함께 사용되던 패키지들과의 충돌을 야기할 수 있기 때문에, 주의해야 한다.

```
$ sudo apt-get upgrade
```

그러면 이제 진짜 패키지를 한번 설치해보자. 첫 번째로 `cmatrix` 라는 패키지를 설치해 보자.

```
$ sudo apt-get install -y cmatrix
```

`-y` 라는 옵션은 설치 중간중간 나오는 질문들에 대해 모두 yes 로 답하겠다는 옵션이다. `-y` 를 빼고 입력해서 중간 중간 직접 yes 를 입력해줘도 된다.

모두 설치가 되었다면 바로 실행해보자.

<img src="./image/cmatrix.png" alt="Terminal" width="70%"/>

프로그램을 종료하고 싶다면 `ctrl` + `c` 를 누르면 된다.

만약 패키지를 삭제하고 싶다면 어떻게 해야할까?

```
$ sudo apt-get remove cmatrix
```

간단하게 `remove` 명령어로 삭제할 수 있다.

## 5. 가상환경으로 패키지를 조금 더 유연하게 관리해보자
가상환경이 무엇인지, 그리고 파이썬에서 가상환경을 다루는 방법이 무엇인지 다음 글을 통해 살펴보자.

[파이썬 코딩 도장](https://dojang.io/mod/page/view.php?id=2470)

가상환경은 프로젝트마다 특정 패키지의 서로 다른 버전이 필요하거나 패키지 간 충돌이 생길 위험이 있을 때, 각 프로젝트를 독립된 공간에서 사용할 수 있도록 하기 위한 기능이다. 충돌이 생기기 전에, 애초에 프로젝트 간 공간을 나누어 놓겠다는 의도를 담고 있다.

여기서 잠깐, 앞으로 명확하게 이해해야 할 용어들을 정리하고 지나가자.

* __패키지(package)__ 특정 기능을 위한 여러 함수 또는 클래스를 담고 있는 보따리로서, 라이브러리(library), 모듈(module)과 비슷한 개념 (다만 모듈은 조금 더 작은 개념으로 쓰이기도 함)

* __가상환경(virtual environment)__ 컴퓨터에 설치된 패키지 간의 충돌 또는 패키지 버전에 의한 이슈 등을 방지하기 위해 가상으로 나누어서 사용하는 환경 즉, 특정 프로그램을 돌리기 위해 필요한 패키지들을 모아 만든 각각의 독립된 방과 같은 개념

파이썬 가상환경을 관리하는 툴은 pyenv, pip, anaconda 등 다양하다. 이 중 __아나콘다(Anaconda)__ 가상환경 관리자를 사용할 것이다. 아나콘다는 데이터 사이언스와 관련된 여러 라이브러리를 포함하여 사용하기 편리하다. 아나콘다를 설치하는 것부터, 가상환경을 생성 및 사용까지 해보자.

### (1) 아나콘다 설치
---

```
$ which conda
/home/username/anaconda3/bin/conda
```

여기서 `which` 명령어는 프로그램 설치 경로를 확인하는 것이다. 윈도우에서는 `where`를 사용한다.

아나콘다를 설치하는 것은 어렵지 않다. 다음 글을 읽어보고 단계별로 따라하자.

[Ubuntu 18.04 - Anaconda 설치 및 가상환경 구축](https://greedywyatt.tistory.com/107)

중요한 것은 우분투 환경을 사용 중이기 때문에 Linux 버전을 다운받아야 한다는 점이다.

글에서 소개했듯 아나콘다 홈페이지에서 리눅스 버전의 설치 파일을 다운받았다면, 다음과 같이 폴더를 이동한 후 설치해준다.

설치파일은 `Downloads` 폴더 내에 다운로드되었을 테니, 그 폴더 내에 아나콘다 설치 파일이 있는지 `ls` 명령어로 확인해보자.

```
$ cd ~/Downloads
$ ls 
Anaconda3-2019.03-Linux-x86_64.sh
```

설치파일이 있다면 `bash` 명령어 뒤에 파일 이름을 작성해서 설치를 진행한다.

```
$ bash Anaconda3-2019.03-Linux-x86_64.sh
```

설치가 완료되면 `.bashrc` 라는 설정 파일에 아나콘다의 환경변수가 등록되어 있을 것이다. 환경변수란, 간단히 말해서 `conda` 라는 명령어를 터미널에서 사용할 수 있도록 만들어주는 것이다. 환경변수가 등록되어 있지 않다면 터미널은 `conda` 라는 명령어를 알아듣지 못한다.

하지만 방금 설치된 아나콘다를 설치한 터미널에는 방금 설정파일에 등록된 `conda`라는 명령어가 적용되어 있지 않을 것이다. `which conda`를 해보면 알 수 있다. 다음 명령어로 `.bashrc` 파일을 한 번 실행해 환경변수를 적용시켜보자.

```
$ source ~/.bashrc
```

하지만 앞으로 새롭게 실행된 터미널에서 굳이 `source ~/.bashrc`를 매번 실행할 필요는 없다. 터미널이 처음 실행될 때마다 설정 파일의 환경변수를 새롭게 읽어들이기 때문이다.

이제 환경변수까지 완료되었으니 `conda`를 실행해보자. 다음 명령어를 통해 설치된 아나콘다의 버전을 확인할 수 있다.

```
$ conda --version
conda 4.8.2
```

### (2) 가상환경 생성 및 실행
---
아나콘다를 설치했다면 다음 글을 통해 가상환경을 직접 만들어보자.

[Create virtual environments for python with conda](https://uoa-eresearch.github.io/eresearch-cookbook/recipe/2014/11/20/conda/)

```
$ conda create -n my_env_name python=3.7
```

`-n` 뒤에 `my_env_name` 과 같이 원하는 가상환경의 이름을 입력하면 된다. `python=3.7` 은 이 가상환경에서는 3.7 버전의 python을 사용하겠다는 의미이다. 가상환경이 잘 만들어졌는지 가상환경 리스트를 확인해보자.

```bash
$ conda env list
# conda environments:
#
base                          *  /home/username/anaconda3
aiffel                              /home/username/anaconda3/envs/aiffel
my_env_name              /home/username/anaconda3/envs/my_env_name
```

가상환경 리스트에 방금 만든 환경이 나타났다면 성공이다 `*` 는 현재 가상환경으로 어떤 환경을 사용하고 있는지 나타낸다.

가상환경을 사용하려면 만든 후에 `activate` 을 해줘야 한다. `activate` 는 문을 열고 그 공간에 들어가는 개념으로 생각하면 된다.

```bash
$ conda activate my_env_name
$ conda env list
base                              /home/username/anaconda3
aiffel                              /home/username/anaconda3/envs/aiffel
my_env_name         *  /home/username/anaconda3/envs/my_env_name
```

다시 리스트를 확인해보면 만든 가상환경이 활성화 된 것을 확인할 수 있다. 이제 새로운 가상환경에 필요한 라이브러리들을 이것저것 설치해서 이 공간을 꾸밀 수 있다.

먼저 이 가상환경 내에 기존에 설치되어 있는 라이브러리 리스트를 확인해보자.

```

$ pip list
Package                            Version
---------------------------------- ----------
alabaster                          0.7.11
anaconda-client              1.7.2
anaconda-navigator       1.9.2
anaconda-project            0.8.2
...
xlwt                                    1.3.0
zict                                      0.1.3
zipp                                     2.2.0
```

여러 가지 라이브러리가 이미 설치되어 있다. 이 가상환경 내에 텐서플로우(TensorFlow)도 설치되어 있을까? 검색해보자.

```
$ pip list | grep tensorflow
```

위에서 사용해봤던 것처럼, `grep` 명령어를 사용하면 원하는 키워드를 검색할 수 있다.

그렇다면 새로운 가상환경에 텐서플로우를 설치해보자. 텐서플로우를 원하는 버전으로 설치하고 싶다면 다음과 같이 버전을 지정해 입력해주면 된다.

```
$ pip install tensorflow==2.2.0
```

> (참고) GPU 버전의 텐서플로우를 설치하려면 `pip install tensorflow-gpu`로 지정해 주어야한다.

모두 설치한 후 다시 리스트를 확인해보자.

```
$ pip list | grep tensorflow

tensorflow           2.2.0
```

리스트에서 tensorflow가 나타난다면 정상 설치 된 것이다. 이렇게 각각의 가상환경을 만들어서 패키지의 종류와 버전을 관리할 수 있다.

하나의 가상환경을 종료하고 싶다면, deactivate 명령어를 사용하면 된다.

```
$ conda deactivate
```

가상환경을 잘못 만들었다면 다음 명령어로 삭제할 수 있다.

```
$ conda env remove -n my_env_name
```

이 외에도 아나콘다를 다루는 다양한 명령어는 다음과 같은 cheatsheet에서 확인할 수 있다.

[Conda Cheat Sheet](https://docs.conda.io/projects/conda/en/4.6.0/_downloads/52a95608c49671267e40c689e0bc00ca/conda-cheatsheet.pdf)

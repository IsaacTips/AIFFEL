## 3. 개발자를 위한 첫 번째 필수 교양 (2) 개발자의 가드닝, 잔디 심기
잔디 심는 개발자라는 말을 들어보았을 것이다. 바로 다음과 같은 잔디밭말이다.

<img src="./image/github.png" alt="github 잔디" width="70%"/>

위 이미지는 바로 GitHub이라는 사이트의 활동 기록이다.

바로 위처럼 나타나는 잔디로 인해, 개발자에게 GitHub은 아주 강력한 버전 관리 호스팅 사이트이면서, 개발자로서 나의 열정을 가장 빠르게 보여줄 수 있는 플랫폼이기도 하다.

위에서 보이는 작은 사각형은 하루하루를 나타낸다. 왼쪽 끝에서부터 오른쪽 끝까지 각 열마다 일주일을 나타내며, 총 365개의 셀로 1년을 보여준다. `commit` 이라는 활동을 해서 GitHub 사이트에 그 내역을 전송하면, 오늘을 나타내는 셀이 초록색으로 칠해진다. `commit` 을 많이 하면 더 진한 초록색으로, 조금 하면 연한 초록색으로 보인다. 아무 활동도 하지 않은 날은 그냥 회색 셀로 보인다.

즉, 위와 같이 모든 셀이 초록색으로 칠해졌다는 것은 1년 365일을 빼먹지 않고 `commit` 했다는 것을 나타낸다. `commit` 을 한다는 것은 어떤 식으로든 코드를 다루었거나 개발과 관련된 문서 작업 등을 했다는 뜻인데, 1년 동안 개발 관련 작업을 단 하루도 빼먹지 않고 했다는 것을 뜻한다. 실제로 저렇게 초록초록한 잔디밭을 만드는 것은 절대 쉬운 일이 아니다.

그만큼 개발자 사이에서는 "1일 1커밋하기" 같은 운동이 있을 정도다. 1일 1커밋을 꾸준히 지속한다면 위와 같이 예쁜 잔디밭을 만들 수 있기 때문에, 저것만으로도 자기 자신이 굉장히 성실한 사람이라는 것을 나타낼 수 있기 때문이다.

하지만 사실 잔디심기는 GitHub의 부수적인 부분일 뿐이다. 실제로 GitHub의 용도는 많은 개발자들이 코드를 오픈소스로 공유하고, 여러 명이 한 팀을 이루어 개발을 진행할 때 복잡한 코드 버전들을 효율적으로 관리하기 위한 호스팅 사이트이다.

## 4. 내 코드의 모든 발자취를 기록할 수 있을까? (1) 버전관리와 Git
슈퍼마리오 게임을 생각해보자.

<img src="./image/mario.png" alt="mario" width="50%"/>

슈퍼마리오 같은 게임에선 클리어해야 하는 맵들이 순서대로 있고, 플레이어는 각 맵에서 여러 번 죽어보기도 하면서 하나하나 게임을 깨나간다. 만약 World 2의 stage 4를 막 클리어했다면, 많은 경우 지금 게임기를 끄고 다음에 다시 키더라도 stage 4의 다음 맵인 stage 5부터 진행할 수 있다. 처음부터 다시 시작할 필요가 없는 것이다.

간단히 말하자면 바로 이런 것이 버전 관리의 대표적인 예라고 할 수 있다. 특정 시점의 진행 상황을 저장해두고, 언제 다시 돌아오더라도 그 시점으로부터 다시 시작할 수 있도록 관리하는 것이다. 만약 되돌아가는 기능도 있다면, 언제든 원하는 과거 시점으로 돌아갈 수도 있을 것이다.

개발을 하면서 코드를 짜다보면 이러한 버전 관리가 필수적이다. 코드를 삭제하거나 덧붙인 다음, 되돌리고 싶을 때 영영 돌아갈 수 없는 상황은 너무 끔찍하다. 아주 간단하고 짧은 코드만을 짠다면 필요가 없겠지만, 코드가 점점 길어지거나 파일이 많아짐에 따라 특정 시점에서의 버전을 잘 기록해두고 관리해야할 필요성이 커진다. 나 혼자서 하는 작업이 아니라, 다른 사람과 함께 한다면 더더욱 그리하다.

소스 코드의 버전 관리를 도와주는 시스템 중 하나가 바로 Git과 GitHub이다. 코드의 버전을 관리함과 동시에 다른 사람과 협업하는 것을 편리하게 만들어주는, 개발자에겐 산소와 같은 프로그램이다.

## 5. 코드의 모든 발자취를 기록할 수 있을까? (2) Git과 GitHub
Git과 GitHub, 이 두 가지의 차이는 무엇일까? 두 용어를 정의내려보자면 다음과 같다.

__Git__

* 개발을 진행하며 작성하는 소스코드가 업데이트 되는 버전을 기록해두고 관리할 수 있는 소스코드 버전 관리 시스템

__GitHub__

* Git으로 관리하는 프로젝트를 호스팅하고, 시간과 공간의 제약 없이 협업할 수 있는 온라인 서비스

이 두 가지를 조금 더 풀어서 설명하자면 다음과 같이 말할 수 있다.

* Git이 버전 기록을 저장한다면, GitHub에서는 그 기록을 다른 사람과 함께 공유하며 협업할 수 있다.
* 로컬(Local)에서 작업한 내용을 Git이 저장해 두었다면, 그 기록을 온라인 작업공간인 GitHub에 올려 원격(Remote)으로도 작업할 수 있도록 한다.

여기에서 로컬이란 개인 노트북 또는 데스크탑같은 Personal Computer를 뜻하고, 원격(Remote)이란 웹사이트와 같이 다른 사람과 함께 작업할 수 있는 공간을 뜻한다.

사실 Git은 리눅스의 창시자 리누스 토르발즈가 만든 오픈소스 툴이다. 전세계 사람들이 오픈스인 리눅스를 함께 개발하고 관리하다 보니 여러 명의 개발자가 짠 코드를 일목요연하게 합치고, 각 버전을 나눌 필요가 생겼다. 그래서 리누스 토르발즈는 리눅스 프로젝트의 코드 버전을 관리하기 위한 소프트웨어를 짜기 시작했는데, 그게 바로 Git인 것이다.

GitHub은 Git이라는 도구를 더 쉽게 사용하게 해주는 사설 서비스이다. 사실 GitHub과 같이 웹사이트 기반으로 Git을 관리하는 온라인 서비스는 GitHub 외에도 GitLab 등 다양하다.

`Ctrl + Alt + T`,  이 단축키로 터미널을 열어보자.

터미널에서 Git을 설치하는 방법은 다음 페이지에서 확인해볼 수 있다.

[우분투 git 설치](https://zetawiki.com/wiki/%EC%9A%B0%EB%B6%84%ED%88%AC_git_%EC%84%A4%EC%B9%98)

아무튼, 터미널을 켰다면 위 페이지에서 알려준대로 다음 명령어로 Git의 설치 여부를 확인할 수 있다.

```bash
$ git --version
```

git 이 설치되어 있지 않다는 내용이 출력되었다면 다음 명령어로 간단히 설치할 수 있다.

```bash
$ sudo apt-get install git
```

```bash
$ git --version
git version 1.9.1
```

## 7. GitHub에 첫 번째 잔디 심기, 어렵지 않아요! (1) 로컬 저장소

### 가. 로컬의 Git에 GitHub의 계정 정보 등록하기
첫 번째로 해야 할 일은 로컬의 Git과 원격에 있는 GitHub을 연결하는 것이다.

Git이 로컬에서 버전 관리를 하는 툴이라면, GitHub은 원격으로 관리하며 협업할 수 있는 웹사이트라고 소개했었다.

이 때 로컬의 Git과 동기화를 해서 온라인으로 관리할 수 있는 원격저장소를 GitHub에서는 레파지토리(Repository)라고 부른다.

따라서 우리가 로컬에서 다양한 코드 작업을 한 후, GitHub의 내 계정에 있는 원격저장소, 레파지토리로 잘 전송하려면 로컬의 Git이 원격의 GitHub 계정 정보를 알고 있어야할 것이다.

다음과 같은 명령어로 간단히 Git과 GitHub을 연결할 수 있다.

```bash
$ git config --global user.email "my-email@gmail.com"
$ git config --global user.name "my-username"
```

위에서 my-email@gmail.com 과 my-username 부분은 자신의 email 주소와 username으로 변경해서 입력하면 된다. 이렇게 입력해주고 나면 Git 툴이 내가 GitHub 사이트로 코드 정보를 전송할 때 어떤 계정에 있는 레파지토리로 전송해야 하는지 기억한다.

아래와 같이, Git에 등록한 config의 정보를 모두 확인하고 싶으면 다음과 같이 입력해보자.

```bash
$ git config -l
```

### 나. 내 컴퓨터에 로컬 저장소 만들기

그 다음으로는 먼저 내 컴퓨터에 Git으로 관리해 보고 싶은 로컬 저장소를 만들어보겠다. 로컬 저장소라는 용어라고 칭하니 거창해 보이지만, 사실은 특별할 것 없이 우리가 보통 파일을 관리할 때 쓰는 폴더, 즉 디렉토리를 말한다. Git은 디렉토리를 기준으로 그 하위에 있는 모든 폴더와 파일에 대한 버전을 기록할 수 있다.

다음과 같이 사용자 홈 디렉토리로 이동한 후 workplace 라는 이름으로 새로운 폴더를 만들어보자.

이 새로운 폴더에서 작업을 시작할 것입니다.

```bash
$ cd ~
$ mkdir workplace
```

### 다. Git으로 버전 관리 시작하기
---
작업을 진행할 새로운 디렉토리 workplace 를 생성하였다.

이제 그 디렉토리로 이동해보자.

```bash
$ cd workplace
```

이제 이 디렉토리를 Git으로 관리하기 시작할 것이다. 다음 명령어로 Git을 현재의 디렉토리 내에 심어놓을 수 있다

```bash
$ git init
```

`init` 은 initialization의 약자이다. 이제부터는 Git이 지금 있는 `workplace` 디렉토리에서 발생하는 모든 변화를 기록해둘 것이다. `ls`만 해서는 보이지 않겠지만 다음과 같이 하면 빈 디렉토리였던 `workspace` 안에 새롭게 생긴 `.git` 디렉토리와, 그 안에 있는 내용들을 확인해 볼 수 있다.

## 8. GitHub에 첫 번째 잔디 심기, 어렵지 않아요! (2) 변화 추적하기

### 라. `README.md` 파일 생성하기
---
README 파일은, GitHub의 레파지토리에서 대문과 같은 역할을 하는 파일이다.

다음의 Tensorflow 오픈소스 코드가 올라와있는 GitHub 레파지토리를 구경하러 가보자.

[tensorflow/tensorflow](https://github.com/tensorflow/tensorflow)

다음 이미지와 같은 README.md 파일을 볼 수 있다.

<img src="./image/readme.png" alt="README" width="50%"/>

이렇게 README.md 파일은 레파지토리를 들어갔을 때 그 레파지토리가 담은 오픈소스 코드들에 대해 소개하는 역할을 한다.

`md` 라는 확장자는 Markdown, 마크다운이라는 파일을 지칭하는데 마크다운이란 개발자들이 가장 많이 사용하는 문서작업 용 언어이다.

이렇게 내 레파지토리를 구경 올 사람들에게 내 작업물에 대해 간단히 소개하기 위한 파일이 바로 `README.md` 파일인 것이다.

그렇다면 첫 번째 레파지토리에 담길 첫 번째 리드미 파일을 만들어보자

```bash
$ echo "# first-repository" >> README.md
```

위와 같은 명령어를 입력하면 `README.md` 파일을 생성함과 동시에 그 파일 내에 `# first-repository` 라는 한 줄이 입력되게 된다.

`echo`는 출력을 하는 명령어인데, 출력 스트림을 지정하는 `>>`을 통해 출력 타겟을 `README.md` 파일로 지정했기 때문이다.

위 명령어를 실행했다면 다음과 같이 `README.md` 파일이 생성된 것을 확인할 수 있을 것이다. `cat` 명령어를 통해 생성된 텍스트 파일을 열어 보자.

* 참고: [Linux manual - cat(1)](https://man7.org/linux/man-pages/man1/cat.1.html)

```bash
$ ls
README.md

$ cat README.md
# first-repository
```

`ls` 로 파일 목록을 확인하면 [README.md](https://readme.md/) 파일이 목록 내에 출력될 것이고, `cat` 명령어를 통해 해당 파일에 작성되어 있는 내용을 확인하면 우리가 입력한 `# first-repository` 가 출력될 것이다.

### 마. git으로 변화 확인! 지금 버전에 도장 쾅! 찍기
---
자, 우리는 방금 git으로 버전을 관리하고 있는 `workplace` 디렉토리에 `README.md` 파일을 생성함으로써 변화를 주었다.

그 변화가 아무리 미미할지언정 변화는 변화이니, git은 그 변화를 감지했을 것이다.

Git이 추적하고 있는 변화는 다음 명령어로 확인할 수 있다.

```bash
$ git status
On branch master

No commits yet

Untracked files:
    (use "git add <file>..." to include in what will be committed)
    README.md

nothing added to commit but untracked files present (use "git add" to track) 
```

위와 같은 내용이 출력되었다면 성공이다. `Untracked files:` 라고 하며 아직 추적되지 않은 새로운 파일인 [`README.md`](http://readme.md) 파일을 잡아내었다.

그렇다면 git에 이 변경사항을 저장해 둘 필요가 있겠다. `add` 와 `commit` 두 가지의 명령어로 진행할 수 있다.

```bash
$ git add README.md
$ git commit -m “new readme file”
[master (root-commit) 438a37c] new readme file 
    1 file changed, 1 insertion(+)
    create mode 100644 README.md
```

`-m`은 메세지 옵션이다. `git commit -m` 뒤에는 해당 커밋에 대한 설명을 작성하면 된다.

`add` 와 `commit` 의 개념은 Git을 다룰 때 아주 중요한 개념 중 하나이다. 두 명령어 모두 현재의 변화를 기록하기 위한 명령어인데, 중요한 차이가 있다.

다음 두 가지 글을 먼저 읽어보고 `add` 와 `commit` 의 차이를 한 번 생각해보자.

[Git 저장소 생성 및 커밋 ( init / add / commit )](https://ifuwanna.tistory.com/193)

[git - 간편 안내서](https://rogerdudler.github.io/git-guide/index.ko.html)

* add는 변화를 기록하기 위한 준비 단계에 해당한다. 파일을 add 하는 것은 staging 한다, 또는 stage에 올려둔다는 등의 표현을 사용하며, 
본격적인 스냅샷(snapshot)을 찍기 전에 임시로 올려두는 개념의 작업이다.

* commit은 실제로 특정 순간의 버전을 스냅샷으로 확정시켜 남겨두는 역할을 한다.

### 사. 내 로컬 저장소와 원격 저장소를 연결해 보자!
---
이제 로컬 저장소도, 원격 저장소도 모두 준비가 되었으니 둘을 연결하는 일만 남았다.

둘을 연결하는 연결고리는 바로 위의 이미지에서 봤던 내 레파지토리의 주소이다.

위 이미지에서 `HTTPS` 라고 적힌 오른쪽에 `https://...` 라고 적혀있는 레파지토리의 주소가 보인다. 저 주소를 로컬 저장소에 있는 git에게 알려주면 git은 GitHub의 원격 저장소의 주소를 알고 두 저장소 간에 정보를 전송할 수 있다. 주소 칸의 맨 오른쪽에 보이는 아이콘을 누르면 주소가 클립보드에 복사된다.

주소를 클립보드에 저장한 후, 터미널에서 아래 명령어를 입력해서 로컬 저장소에 있는 Git이 원격 저장소의 주소를 알 수 있도록 저장하자. 아래 명령어는 꼭 로컬 저장소 디렉토리 안에서 실행해야 한다.

```bash
$ cd ~/workplace
$ git remote add origin https://github.com/xxx/first-repository.git
```

`origin` 은 원격 저장소의 닉네임과 같은 역할을 한다. 앞으로는 복잡한 `https://...` 의 주소를 매번 사용하는 것이 아니라, `origin` 이라는 이름으로 원격 저장소를 지칭할 수 있는 것이다. 물론 위의 `xxx` 부분은 본인의 username이 들어가야한다

### 아. 로컬 저장소의 기록을 원격 저장소로 전송하기
---
이제 모든 준비가 끝났으니 로컬에서 레파지토리로 전송하는 것은 아주 간단하다.

바로 다음 명령어 한 줄이면 된다.

```bash
$ git push origin master
Enumerating objects: 3, done.
Counting objects: 100% (3/3), done.
Writing objects: 100% (3/3), 230 bytes | 230.00 KiB/s, done.
Total 3 (delta 0), reused 0 (delta 0)
To https://github.com/jeina7/first-repository.git
    * [new branch]      master -> master
```
    
위 명령어는 현재 로컬에 있는 버전 기록과 모든 파일들을 `origin`, 즉 원격 저장소의 master 브랜치로 `push`해 밀어넣겠다는 뜻을 가진다.

브랜치란, 간단히 말하자면 작업을 하는 '가지'를 말한다. 하나의 작업을 여러 개의 가지로 분기하여 작업을 진행할 수 있다.

브랜치의 개념은 다음 글에서 참조.

[브랜치란? 【브랜치 (Branch)】 Backlog](https://backlog.com/git-tutorial/kr/stepup/stepup1_1.html)

* 브랜치는 여러 작업을 독립적으로 진행하려고 할 때 필요하다. 필요에 의해 만들어지는 각각의 브랜치는 다른 브랜치의 영향을 받지 않기 때문에, 여러 작업을 동시에 진행할 수 있다.

### 자. 원격 저장소를 로컬로도 가져올 수 있을까?
---
만약 지금까지 했던 것처럼 로컬 저장소를 GitHub의 원격 저장소로 전송하는 것이 아니라, 그 반대로 이미 GitHub에 올라와 있는 저장소를 통째로 내 로컬에 가져오고 싶다면 어떻게 해야 할까?

이 작업 또한 아주 간단하다.

원격 저장소를 로컬에 가져오려면 역시 그 연결고리가 필요하다. 다음과 같이 레파지토리의 주소를 복사해보자.

<img src="./image/first_repo3.png" alt="first-repository" />

우리는 이 저장소를 아까 생성했던 `workplace`가 아닌 다른 디렉토리인 `project`에 가져와보자.

```bash
$ cd ~
$ mkdir project
$ cd project
```

이동 후에, 다음과 같은 명령어를 사용하면 GitHub의 레파지토리를 통째로 가져올 수 있다. 명령어는 복제라는 뜻을 가진 `clone` 을 사용한다.

```bash
$ git clone https://github.com/xxx/first-repository.git
Cloning into 'first-repository'...
remote: Enumerating objects: 3, done.
remote: Counting objects: 100% (3/3), done.
remote: Total 3 (delta 0), reused 3 (delta 0), pack-reused 0
Unpacking objects: 100% (3/3), done.
```

그러면 제대로 잘 복사가 되었는지 보자.

```bash
$ ls
first-repository
```

`ls` 로 확인하면 레파지토리 이름으로 폴더가 생성된 것을 확인할 수 있다. 이동해서 확인해보자.

```bash
$ cd first-repository
$ ls
README.md
```

폴더 내로 이동해서 다시 `ls` 로 확인하면 아까 만들어놓고 레파지토리로 전송했던 `README.md` 파일을 확인할 수 있다. `cat` 으로 내용을 확인해보자.

```bash
$ cat README.md
# first-repository
```

내용도 작성한대로 잘 가져와졌다.

### 차. 로컬로 가져온 원격 저장소를 수정해서 다시 push 해 보자!
---
그렇다면 이제 로컬로 가져온 레파지토리 내용을 수정해서 다시 원격으로 전송해 보는 작업을 할 것이다.

파일에 변화를 주기 위해서 현재 있는 `README.md` 파일을 한번 수정해보자.

```bash
$ echo "add new contents" >> README.md
```

이번에도 `echo` 명령어를 활용하지만, 아까와는 다르게 이미 `README.md` 파일이 있기 때문에 새로 생성하는 것이 아니라 현재 있는 파일의 아래쪽에 문자열을 추가하게 된다.

다시 한 번 파일 내용을 확인해보자.

```bash
$ cat README.md
# first-repository
add new contents
```

내용에 원래 있던 첫 번째 줄과 아래 두 번째 줄이 추가되었다.

그렇다면 다시 변화가 생긴 것이니, git이 이 변화를 추적하고 있을 것이다. 확인해보자.

```bash
$ git status
On branch master
Changes not staged for commit:
    (use "git add <file>..." to update what will be committed)
    (use "git restore <file>..." to discard changes in working directory)
    modified:   README.md

no changes added to commit (use "git add" and/or "git commit -a")
```

이번에는 [`README.md`](http://readme.md) 파일이 새로 생긴 것이 아니라, `modified` 되었다고 나온다. 맞게 잘 추적하고 있는 것 같다.

이제 다시 아까 했던 작업과 동일하게 `add`, `commit` 을 진행하도록 하자.

```bash
$ git add README.md
$ git commit -m “new contents”
[master c82640d] new contents
    1 file changed, 1 insertion(+)
```

새로운 `commit`, 즉 새로운 스냅샷이 저장되었다.

마지막 단계는 이것을 다시! 원격 저장소로 전송하는 것이다. 원격 저장소로 보내는 명령어는 `push`였다!

```bash
$ git push origin master
Enumerating objects: 5, done.
Counting objects: 100% (5/5), done.
Writing objects: 100% (3/3), 276 bytes | 276.00 KiB/s, done.
Total 3 (delta 0), reused 0 (delta 0)
To https://github.com/jeina7/first-repository.git
    438a37c..c82640d  master -> master 
```

위와 같은 결과가 나오면 모두 완료된 것이다.

이제 최종적으로 원격 저장소, 레파지토리로 다시 가서 확인해보자.

<img src="./image/first_repo4.png" alt="first-repository" />

새로운 내용이 잘 추가되었다.

### 카. 로컬 저장소를 원격 저장소의 내용과 같게 업데이트하자!

처음에 만들었던 로컬 저장소의 `README.md` 파일이 자동으로 업데이트 되었을까? 다시 이동해서 확인해보자.

```bash
$ cd ~/workplace
$ ls
README.md
$ cat README.md
# first-repository
```

음, 당연하지만 자동으로 업데이트가 되지는 않았다. 하지만 다행히 이 로컬 저장소를 수정된 원격 저장소와 같게 업데이트 해주는 것 또한 매우 간단합니다.

```bash
$ git pull origin master
remote: Enumerating objects: 5, done.
remote: Counting objects: 100% (5/5), done.
remote: Total 3 (delta 0), reused 3 (delta 0), pack-reused 0
Unpacking objects: 100% (3/3), done.
From https://github.com/jeina7/first-repository
    * branch            master     -> FETCH_HEAD
    438a37c..c82640d  master     -> origin/master
Updating 438a37c..c82640d
Fast-forward
    README.md | 1 +
    1 file changed, 1 insertion(+)
```    

`push` 의 반댓말, `pull` 로 `origin` 별칭의 원격 저장소를 로컬 저장소로 "당겨올" 수 있다.

잘 가져와졌다면, 다음과 같은 결과가 나와야한다.

```bash
$ cat README.md
# first-repository
add new contents
```

`add new contents` 라는 내용이 추가되었다면 성공이다!

## 12. GitHub에 첫 번째 잔디 심기, 어렵지 않아요! (6) 정리

여기까지 모두 진행했다면, 우리의 GitHub 프로필 페이지에서 초록색 칸으로 색칠이 된 잔디를 확인 할 수 있을 것이다.

이렇게 잔디가 칠해지려면 `commit`, `push` 등의 GitHub 활동을 해야한다. 즉, 맨 처음에 봤던 그 1년 내내 파릇파릇한 잔디는 이러한 코드 작업을 매일매일 한 것으로 볼 수 있다.

지금까지 무엇을 한 걸까?

그림으로 간단히 정리하자면 다음과 같다.

<img src="./image/git.png" alt="git" />

`workplace` 라는 로컬저장소, 다른 말로는 디렉토리를 만드는 것부터 시작해서, `add`, `commit`, `push 세 가지의 명령어를 통해 원격 저장소로 전송하는 것까지 진행해 보았다.

그 후에는 원격 저장소를 다시 `first-repository` 라는 디렉토리로 복제해와서 파일을 수정한 후 다시 원격저장소로 전송했다.

<img src="./image/git2.png" alt="git" />

위에서 봤던 `workplace` 를 만든 것이 개발자 A, 원격 저장소를 복제해왔던 `first-repository` 를 다루는 사람이 개발자 B라고 해보자.

협업을 진행하는 두 개발자가 GitHub을 활용한다면, 둘은 서로 업데이트한 __파일을 따로 전송할 필요 없이__ GitHub에 Push 또는 Pull 하는 것만으로 효율적인 협업을 진행할 수 있는 것이다. 지금은 크게 중요하지 않아 보일 수 있지만, 여러 개발자가 동시에 개발을 진행하며, 여러 개의 파일에 작업을 하고, 그 버전들이 다양해 질 경우, 각 파일의 버전이 관리되지 않는다면 파일을 관리하는 작업은 굉장히 복잡해질 수 있다.

이러한 어려움을 Git과 GitHub은 `commit` 이라는 버전 관리 작업을 통해 손쉽게 관리할 수 있도록 한 것이다.

### 3. 노트북 마음대로 사용하기, 어렵지 않습니다!
---
위와 같은 빈 노트북 화면이 나타났다면 이제 노트북의 사용법을 알아보자.

<img src="./image/jupyter7.png" alt="Jupyter Notebook" />

노트북에는 기본적으로 위와 같은 초록색의 입력 모드와, 파란색의 명령 모드가 있다.

처음으로 노트북을 실행하면 기본적으로 입력 모드로 시작하게 되며, `esc` 키를 누르면 명령 모드로 바뀐다. 입력 모드일 때에는 말 그대로 셀에 코드나 문자를 입력할 수 있고, 명령 모드일 때에는 셀에 대한 설정을 바꾸거나 옮기기, 또는 새로운 셀 추가 등의 작업을 할 수 있다.

<img src="./image/jupyter8.png" alt="Jupyter Notebook" />

다음으로 셀의 타입으로는 위와 같이 두 가지가 있다. 마크다운 셀은 제목 또는 설명 등을 입력하고, 코드 셀은 실행시킬 수 있는 파이썬 코드를 입력한다.

특정 셀을 마크다운 셀 또는 코드 셀로 변환하고 싶다면 다음과 같은 단축키를 사용한다.

* 마크다운 셀로 변환하기 : `esc` + `m` (명령모드로 변환 후 `m` )
* 코드 셀로 변환하기 : `esc` + `y` (명령모드로 변환 후 `y` )

이 외에 가장 많이 쓰이는 노트북 단축키는 다음과 같다. 단축키는 정말 살이 되고 피가 되니, 한 번씩 누르면서 익혀보자.

* 셀의 실행 : `Shift` + `Enter`
* 셀 삭제 : `esc` + `x` or `esc` + `dd`
* 셀 삭제 취소 : `esc` + `z`
* 위에 셀 추가 : `esc` + `a`
* 아래에 셀 추가 : `esc` + `b`

단축키 사용 시 가장 많이 쓰이는 키는 바로 `esc` 키이다. 특정 행동을 하려고 할 때 주로 명령 모드에서 많이 진행하기 때문이다.

이 외에도 주피터 노트북에는 단축키가 많다, 다른 단축키가 궁금하다면 명령모드에서 `H` 를 눌러보자.

다음과 같은 단축키 모음을 확인할 수 있다.

<img src="./image/jupyter9.png" alt="Jupyter Notebook" />

## 16. 개발자 문서작업의 시작과 끝! 마크다운을 익혀보자 (2) 마크다운 작성하기

### (1) 제목 달기
---
마크다운에서 가장 많이 쓰이는 것은 아마 제목일 것이다. 제목은 `#` 으로 달 수 있으며, `#` 의 개수에 따라 제목의 크기가 달라진다.

# This is a H1
## This is a H2
### This is a H3

```
# This is a H1
## This is a H2
### This is a H3
#### This is a H4
##### This is a H5
###### This is a H6
```

위와 같이 `#` 이 많아질수록 작은 제목이 되며, 마크다운에서는 H6까지 지원하지만 어떤 툴로 마크다운을 사용하냐에 따라 H4~H6 까지는 지원이 안 될 수도 있다.

3. 세번째

```
1. 첫번째
2. 두번째
3. 세번째
```

### (5) 볼드체, 이태리체, 취소선 등 강조하기
---
글씨 양쪽에 `*` 또는 `_` 로 묶어주면 볼드체 또는 이태리체를, `~` 로 묶어주면 취소선을 나타낼 수 있다. (취소선은 브라우저 또는 환경에 따라 다르게 나타날 수 있다)

*single asterisks*<br>
_single underscores_<br>
**double asterisks**<br>
__double underscores__<br>

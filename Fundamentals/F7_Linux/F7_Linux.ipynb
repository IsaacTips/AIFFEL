```python
june@juneoh-ubuntu:~$ pwd
/home/june
```

> 👍️ **`pwd`**: Present Working Directory의 약자이다. 현재 내가 위치한 디렉토리 위치를 출력한다.

**`/home/사용자명`**과 같은 형태의 결과가 표시된다. **`home`** 폴더 안의 **`사용자명`** 폴더 안에 있다는 뜻인데, 윈도우즈에서는 폴더 구분에 역슬래시(**`\`**, 또는 글꼴에 따라 원화 기호 **`₩`**로 출력)를 쓰지만 리눅스에서는 슬래시(**`/`**)를 쓰는 것을 알 수 있다.

이 위치를 해당 사용자의 홈 디렉토리라고 한다. 사용자마다 하나씩 주어지며, 윈도우즈의 내 문서 폴더와 비슷하다고 생각하면 된다. 홈 디렉토리를 줄여서 **`~`**로도 표시하는데, 이는 옛날 옛적에 Home 키와 ~ 키가 같은 자판에 있던 시절로부터 전해져 내려온 유구한 전통이다.

그럼 이제 아까부터 터미널에 떠 있던 저 문구의 뜻을 조금은 알 수 있을 것 같다.

```bash
june@juneoh-ubuntu:~$
```

* @ 앞의 **`june`**은 현재 로그인한 사용자명이다. 
* @ 뒤의 **`juneoh-ubuntu`**는 컴퓨터의 이름, 호스트명이다.
* : 뒤의 **`~`**는 현재 위치이다. 마지막 **`$`**는 현재 로그인한 사용자가 일반 사용자임을 나타낸다. 나중에 배울 최고 관리자 계정(root)의 경우에는 **`#`**로 표시된다.

리눅스는 UNIX라는 운영체제를 본떠 만들어졌는데, 이 UNIX의 철학 중 하나는 "모든 것은 파일이다"라는 점이다. 윈도우에서는 각종 시스템 설정값이 레지스트리에 들어있지만, 리눅스에서는 설정값 또한 파일이며, 키보드 등의 하드웨어 기기들도 파일로서 접근 가능하다. 셸 내부에서 나의 위치도 마치 하나의 파일처럼 항상 어떤 폴더 안에 있기 때문에 길을 잃을 염려는 없다.

우리가 지금 사용하고 있는 것과 같은 프로그램의 종류를 **셸(shell)** 이라고 하며, 우리는 이 껍데기를 통해 운영체제의 알맹이인 **커널(kernel)** 과 소통한다. 셸이 무엇인지에 대해서는 뒤에서 자세히 알아보겠다.

## 3. 떠나자 디지털 월드로! (2) ls

### **`ls`: 뭐가 있지?**

---

우리는 여기서 어디로 갈 수 있을까? **`ls`**를 입력하여 지금 있는 폴더 안에 어떤 파일/폴더가 있는지 목록을 뽑아보자.

```bash
june@juneoh-ubuntu:~$ ls
Desktop  Documents  Downloads  Music  Pictures  Public  Templates  Videos
```

> ❤️ **`ls`**: 현재 디렉토리 내의 파일 및 폴더들을 출력한다.

하지만 세상에 눈에 보이는 것이 전부는 아니다. **`ls`** 대신 **`ls --all`**를 입력해보자.

```bash
june@juneoh-ubuntu:~$ ls --all
.   .bash_logout  .cache   Documents  .gnupg  Pictures  Public  Templates
..  .bashrc       Desktop  Downloads  Music   .profile  .ssh    Videos
```

아까는 안 보이던 파일들이 늘어났다. 어떻게 된 것일까?

**`~`**가 홈 디렉토리를 가리키는 것처럼, **`.`**는 현재 폴더를, **`..`**는 상위 폴더를 가리킨다.

참고로 리눅스에서 **`.`**으로 시작하는 이름을 가진 파일은 숨김 파일이다. 예를 들어 **`.bashrc`**와 **`.profile`**은 셸에 로그인 시 자동으로 불러오는 설정 파일들이다. 이런 숨김 파일들도 **`ls`** 명령어에 **`--all`** 인자를 함께 사용하면 모두 표시할 수 있다.

그런데 어떤 명령어가 어떤 인자를 제공하는지는 어떻게 할 수 있을까?

대부분의 명령어들은 **`--help`**라는 인자를 기본으로 제공한다. **`ls --help`**를 입력해보자.

```bash
Exit status:
 0  if OK,
 1  if minor problems (e.g., cannot access subdirectory),
 2  if serious trouble (e.g., cannot access command-line argument).

GNU coreutils online help: <http://www.gnu.org/software/coreutils/>
Full documentation at: <http://www.gnu.org/software/coreutils/ls>
or available locally via: info '(coreutils) ls invocation'
june@juneoh-ubuntu:~$
```

내용이 너무 길어서 화면이 넘어갔다. 지금 화면에서는 마우스도 쓸 수 없는데 어떡해야 할까? **`Ctrl+Shift+↑/↓`**으로 한줄씩 위/아래로 스크롤하거나, **`Shift+PageUp/PageDown`**으로 한 페이지씩 위/아래로 스크롤 할 수 있다.

```bash
june@juneoh-ubuntu:~$ ls --help
Usage: ls [OPTION]... [FILE]...
List information about the FILEs (the current directory by default).
Sort entries alphabetically if none of -cftuvSUX nor --sort is specified.

Mandatory arguments to long options are mandatory for short options too.
  -a, --all                  do not ignore entries starting with .
  -A, --almost-all           do not list implied . and ..
      --author               with -l, print the author of each file
```

**`-all`**은 **`a`**로 줄여서 쓸 수도 있다는 점을 알았다. 추가로 다음 인자들도 눈여겨보면 좋다.
- **`l`**: 각 파일 및 폴더의 권한, 소유자, 크기, 수정일시 등 자세한 정보를 출력한다.
- **`h`**: 파일 크기를 바이트 수 대신 사람이 알아볼 수 있는 단위(1K, 234M, 2G 등)로 표시한다.
- **`t`**: 파일 수정시간 순으로 정렬해서 출력한다. 최신 파일이 위쪽으로 표시된다.
- **`r`**: 정렬 순서를 뒤집어 출력한다. **`rt`**의 형태로 쓰면 최신 파일이 아래쪽으로 표시되어 유용할 때가 있다.

한 글자 인자를 여럿 쓸 때에는 서로 붙여 쓸 수 있다. 즉, **`ls -l -a -h`**는 **`ls -lah`**로 쓸 수 있다.

```bash
june@juneoh-ubuntu:~$ ls -lah
total 64K
drwxr-xr-x 13 june june 4.0K Mar 17 01:20 .
drwxr-xr-x  6 root root 4.0K Mar 17 00:15 ..
-rw-r--r--  1 june june  220 Apr  4  2018 .bash_logout
-rw-r--r--  1 june june 3.7K Apr  4  2018 .bashrc
drwx------  2 june june 4.0K Mar 17 00:16 .cache
drwxrwxr-x  2 june june 4.0K Mar 17 01:17 Desktop
drwxrwxr-x  2 june june 4.0K Mar 17 01:20 Documents
drwxrwxr-x  2 june june 4.0K Mar 17 01:17 Downloads
drwx------  3 june june 4.0K Mar 17 00:16 .gnupg
drwxrwxr-x  2 june june 4.0K Mar 17 01:20 Music
drwxrwxr-x  2 june june 4.0K Mar 17 01:20 Pictures
-rw-r--r--  1 june june  807 Apr  4  2018 .profile
drwxrwxr-x  2 june june 4.0K Mar 17 01:20 Public
drwx------  2 june june 4.0K Mar 17 00:15 .ssh
drwxrwxr-x  2 june june 4.0K Mar 17 01:20 Templates
drwxrwxr-x  2 june june 4.0K Mar 17 01:20 Videos
```

(Tip) **`ls -alrt`** 가 유용할 때가 많다.

## 4. 떠나자 디지털 월드로! (3) cd

### **`cd`: 움직여보자**

---

리눅스에서 모든 것은 파일이고, 나도 파일처럼 항상 어느 폴더 안에 들어있다는 사실을 위에서 배웠다. 그럼 내가 다른 폴더로 이동하려면 어떻게 해야할까? 내가 들어있는 디렉토리를 바꾸면 된다. **`cd aiffel`**를 입력해 aiffel 폴더 안으로 이동해보자.

```bash
june@juneoh-ubuntu:~$ cd aiffel
june@juneoh-ubuntu:~/aiffel$
```

> ❤️ **`cd`**: Change Directory의 약자로, 뒤에 따라오는 폴더로 내가 있는 셸의 현재 위치를 이동시킨다.

커서 앞의 **`~`**가 **`~/aiffel`**로 바뀌었다. 여기서 **`pwd`**를 해보면?

```bash
june@juneoh-ubuntu:~/aiffel$ pwd
/home/june/aiffel
```

여기서 다시 상위 폴더로 이동하려면? **`..`**이 상위 폴더를 가리킨다고 했으니, **`cd ..`**를 입력한 뒤 이어서 **`pwd`**를 해보자.

```bash
june@juneoh-ubuntu:~/aiffel$ cd ..
june@juneoh-ubuntu:~$ pwd
/home/june
```

여기서 **`..`**나 **`aiffel`**에서처럼 명령을 입력하는 시점에 내가 현재 있는 디렉토리에 영향을 받는 위치 표시 방식을 상대 경로라고 한다. 반면에 **`/home/june/aiffel`**와 같이 어디서 입력하든 똑같은 곳을 가리키도록 위치 전체를 적어주는 것을 절대 경로라고 한다.

> **뒤에 위치 없이 그냥 `cd`만 입력하면 어디로 갈까?**

그럼 이 세상의 끝은 어디일까? 계속 한번 위로 올라가보자.

```bash
june@juneoh-ubuntu:~$ cd ..
june@juneoh-ubuntu:/home$ pwd
/home
june@juneoh-ubuntu:/home$ cd ..
june@juneoh-ubuntu:/$ pwd
/
june@juneoh-ubuntu:/$ cd ..
june@juneoh-ubuntu:/$ pwd
/
```

**`/`**에 도달한 이후 더 이상 상위 폴더로 올라가지지 않는다. 여기가 바로 세상의 끝이다.

**세상의 끝에는 무엇이 있을까?**

```bash
june@juneoh-ubuntu:/$ ls
bin   etc         initrd.img.old  lost+found  opt   run   srv  usr      vmlinuz.old
boot  home        lib             media       proc  sbin  sys  var
dev   initrd.img  lib64           mnt         root  snap  tmp  vmlinuz
```

일부 주요 폴더들의 용도 및 내용물은 다음과 같다. 굳이 기억할 필요는 없고, 이런 것들이 있다는 것만 알고 넘어가자.

* **`/home`**: 사용자별 홈 디렉토리들이 있는 곳
* **`/root`**: 최고 관리자 계정의 홈 디렉토리
* **`/mnt`**: 저장장치(HDD, SSD)가 붙는 위치
* **`/media`**: 이동식 미디어(USB 드라이브)가 붙는 위치
* **`/tmp`**: 재부팅 시 삭제될 임시 파일들을 저장하는 폴더
* **`/dev`**: 컴퓨터에 연결된 하드웨어 및 가상 기기(device)들을 가리키는 파일들
* **`/proc`**: 현재 실행 중인 프로세스들을 가리키는 파일들
* **`/etc`**: 각종 설정 파일들
* **`/bin`**: 실행 가능한 프로그램(binary)들
* **`/sbin`**: 시스템 관리용 프로그램들
* **`/usr`**: 다중 사용자 모드에서 사용 가능한 파일 및 프로그램들 (root 계정만 있는 단일 사용자 모드에서는 사용 불가)
* **`/var`**: 캐시, 로그 등 시스템 구동 간 계속 내용이 바뀌는 파일들

한번 어디로 들어가 볼까? 최고 관리자 계정이 탐이 난다.

```bash
june@juneoh-ubuntu:/$ cd root
-bash: cd: root: Permission denied
```

보이지 않는 무언가에 가로막혔다. 무슨 권한이 더 필요하다고 하는데, 일단 세계의 끝도 봤으니 지금은 여기서 돌아가보자. **`Ctrl+D`**를 눌러 로그아웃한 뒤, **`Ctrl+Alt+F2`**을 눌러 우리가 알던 세계로 돌아가자.

## 5. 운영체제 일반 개론 (1) 운영체제란

잠시 안도의 숨을 돌리면서, 리눅스를 넘어 일반적인 운영체제에 대해 이야기해보겠다. 운영체제란 컴퓨터의 다양한 하드웨어 장치들과 더불어 소프트웨어를 관리하는 시스템을 일컫는다.

막연하실 것 같아 이런 예를 들어 보겠다.<br>
커다란 프랑스 귀족 저택에 요리사가 3명, 정원사 5명, 청소부가 10명이나 있다고 하자. 심지어 이 하인들은 출신국가나 언어도 제각기 달라서 어떤 요리사는 영어를, 어떤 청소부는 독일어를, 어떤 정원사는 중국어를 쓰기도 한다고 하자. 이 저택의 집주인이 이 많은 하인들에게 일을 직접 시키려면 얼마나 힘들까? 적절하게 필요 자원을 배분해서 일을 시키는 것만으로도 바쁜데, 그 수많은 하인들의 언어와 성격까지 맞추어 일을 시킨다는 것은 보통 일이 아닐 것이다.<br>
이때, 전세계 모든 언어를 마스터한 스마트한 집사가 하나 나타났다고 생각해보자. 이제 이 귀족은 굳이 하인들을 직접 만날 필요가 없다. 그냥 스마트한 집사에게 모든 것을 시키면 된다. 설령 요리사가 도중 아프리카인으로 교체된다고 해도 귀족 집주인은 신경쓸 필요가 없다. 달라지는 부분에 대해서는 집사가 다 알아서 해줄 것이기 때문이다.

비유해서 생각하면 운영체제는 컴퓨터라는 대저택을 관리하는 집사와 같은 존재로 볼 수 있다. 집주인(최고관리자)의 권한을 위임받아 요리사, 정원사, 청소부 등 각각의 보조 인원(어플리케이션)들에게 재산(연산 자원)을 적절히 분배하여 운영하고, 서로 권한을 침범할 수 없도록 관리하는 역할을 맡는다.

<img src="./image/os.png" />

[https://en.wikipedia.org/wiki/Kernel_(operating_system)]
우리가 지금 보고 있는 것은 운영체제들 중 리눅스 계열의 우분투(Ubuntu)라는 운영체제이다. 조금 더 친숙한 운영체제로는 마이크로프트의 윈도우즈나, 애플의 맥 OS, iOS, 구글의 Android, ChromeOS 등이 있다.

위 그림에서 CPU, Memory, SDD 등 각종 디바이스를 통합관리하면서 그 연산자원을 수많은 어플리케이션들에게 분배하는 자리에 있는 것은 바로 커널(Kernel)이라는 것이다. *운영체제=커널*이라고 할수는 없겠지만 커널은 보안, 자원관리, 디바이스 인터페이스 추상화 등 바로 대저택의 집사와 같은 역할을 하는 운영체제의 가장 핵심적인 구성요소이다.

## 6. 운영체제 일반 개론 (2) 커널

운영체제의 내부는 알맹이인 커널(kernel)과 껍데기인 셸(shell)로 이루어져 있다. 우분투의 커널은 리눅스이고, 마이크로소프트 윈도우(Windows)의 경우 2000 이후의 버전은 Windows NT라는 커널을 사용하며, 맥 OS(macOS)의 경우에는 XNU(XNU is Not Unix의 약자)라는 커널을 사용한다.

<img src="./image/win.jpeg" />

__그래서 버전이 8이라는 거야 6.3이라는 거야?__

* 전체 운영체제의 버전이 8이고, 커널의 버전이 6.3이라는 뜻이다.

윈도우를 제외한 운영체제들은 사실상 서로 사촌지간이라고 볼 수 있는데, 여기에는 약간의 역사가 있다. 태초에 유닉스(Unix)라는 상용 운영체제(겸 해당 운영체제를 위한 커널)가 있었다. 이를 기반으로 미국 UC 버클리에서 개발한 BSD 운영체제가 나오자, 미국 카네기 멜론 대학교에서 BSD를 위해 Mach 커널을 개발한다. 1985년 애플에서 쫓겨났던 스티브 잡스가 창업한 NeXT에서는 이 Mach 커널을 사용해 XNU라는 커널을 만든다. 이후 NeXT가 (스티브 잡스와 함께) 애플에 흡수되면서 XNU는 맥 OS, iOS (및 iPadOS, tvOS, watchOS..) 운영체제에서 커널로 사용되게 된다.

리눅스 커널 또한 유닉스와 관련이 깊은데, 1991년 핀란드에서 당시 컴퓨터 공학과 학생 리누스 토발즈(Linus Torvalds)가 유닉스의 무료 및 공개 버전을 지향하여 만든 것이 리눅스이다. 여기서 명심할 점은 리눅스가 그 자체로 운영체제가 아니라, 운영체제를 위한 커널이라는 사실인데, 우리가 흔히 리눅스 운영체제라 부르는 것들은 사실 이 리눅스를 커널로써 사용하는 운영체제들을 가리키며, 우분투나 Android, ChromeOS 말고도 Debian, CentOS, Red Hat Enterprise Linux 등 다양한 종류가 있다.

즉, 맥 OS의 XNU 커널과 우분투의 리눅스 커널은 역사적 배경 때문에 모두 유닉스와 유사하며, 유닉스 표준(POSIX)과 일부 호환성이 제공된다. 이런 유사성은 두 운영체제 모두 동일하게 bash를 기본 셸로 제공하는 점에서도 찾아볼 수 있다.

> bash가 뭘까? 궁금하시면 **`Ctrl+Alt+T`**를 누른 후 **`echo $SHELL`** 이라고 입력해보자.

## 7. 운영체제 일반 개론 (3) 셸

셸은 말 그대로 눈에 보이는 운영체제의 껍데기인데, 그래픽 기반의 GUI(Graphical User Interface) 셸과 텍스트 기반의 CLI(Command Line Interface) 셸로 구분할 수 있다.

**GUI 셸**의 경우, 윈도우에는 우리에게 친숙한 시작 버튼을 포함하는 셸을 Windows shell이 라 부르고, 맥 OS의 경우는 Aqua라는 이름을 갖고 있다. 리눅스 계열 운영체제는 GNOME, Unity(동명의 게임 엔진과는 별도의 프로그램) 등이 있지만, 단독으로 동작하기보다 보통 Xorg 등 다른 종류의 프로그램들과 함께 사용된다.

**CLI 셸**로는 윈도우 운영체제의 cmd.exe와 PowerShell이 있고, 우분투 및 맥 OS에서는 CLI 셸로서 기본으로 sh와 여기에 몇 가지 편의기능을 덧붙인 bash가 설치되어 있다. 이 외에도 csh, fish, zsh 등 다양한 CLI 셸을 직접 설치하여 사용할 수 있다. 또한 zsh처럼 테마 및 플러그인을 지원하는 셸의 경우 모양과 편의기능을 입맛에 따라 꾸밀 수 있다.

## 8. 운영체제 일반 개론 (4) 프로세스와 스레드

**프로그램**은 컴퓨터가 수행할 작업을 기록해놓은 것을 뜻한다. 프로그램이 실제로 작업으로써 수행될 때 운영체제는 이를 **프로세스(process)**라는 단위로 관리한다. 즉, 운영체제는 프로세스 단위로 메모리와 CPU 연산 등 컴퓨팅 자원을 분배한다.

* **메모리 관리**: 운영체제는 프로세스가 컴퓨터의 전체 메모리 어디든지 쓸 수 있도록 하는 것이 아니라 가상 메모리(virtual memory)로써 일부만 떼어서 제공해줌으로써, 다른 프로세스의 메모리 또는 운영체제 자체가 사용하고 있는 커널 메모리를 훔쳐볼 수 없도록 한다. 이것을 운영체제의 중요한 보안 기제 중 하나인 프로세스 격리(process isolation)라고 부르며, 때문에 프로세스들이 서로 소통하기 위해서는 별도로 허용된 프로세스 간 통신 (Inter-Process Communication, IPC) 기법들을 사용해야 한다.
* **CPU 자원 관리**: 메모리는 주소로서 프로세스에게 제공되지만, CPU 연산의 경우 코어 단위로 프로세스에 제공된다. 프로세스는 **스레드(thread)**라는 단위로 코어를 하나씩 사용할 수가 있는데, 프로세스가 하나 시작되면 기본적으로 하나의 스레드를 가지고 시작하지만, 개발자는 프로그램이 더 많은 스레드를 사용하도록 설계하여 다중 CPU 코어의 이점을 극대화할 수 있다.

<img src="./image/thread.png" />
<center>[https://en.wikipedia.org/wiki/Thread_(computing)]</center>

프로세스는 운영체제가 메모리를 할당하는 **작업단위**라면, 스레드는 프로세스가 할당받은 메모리를 활용하는 **실행단위**이다.

하나의 프로그램 안에서 여러개의 프로세스를 활용하는 것을 **멀티프로세싱(Multiprocessing)**, 하나의 프로세스 안에서 여러개의 스레드를 활용하는 것을 **멀티스레딩(Multithreading)**이라고 한다.



## 9. 리눅스 기반 운영체제 (1) 터미널

위에서는 일반적인 운영체제에 공통적인 내용을 다루었지만, 이제부터는 리눅스 기반 운영체제에 조금 더 특화된 내용을 살펴보겠다.

우분투에서 **`Ctrl+Alt+T`**를 눌러보면 아까 본 텍스트 화면이 들어있는, 흔히 터미널 또는 콘솔이라고도 부르는 창이 열린다. 이 창 내부에서 텍스트로 조작하는 환경은 위에서 배웠듯이 CLI 셸이다. 이 창을 여는 프로그램은 엄밀히 말하면 터미널이 아니라, 정확히는 그래픽 환경에서 터미널을 모사하는 **터미널 에뮬레이터**이다. 그렇다면 진짜 터미널이란 무엇을 가리키는 걸까?

터미널이라는 이름은 개인용 컴퓨터(PC)가 등장하기 전의 시절, 실제 연산을 담당하는 크고 비싼 메인프레임 컴퓨터와는 별도로, 입출력만 담당하는 CRT 모니터와 키보드로 구성된 말단(=terminal)의 기기에서 유래한 명칭이다. 여러 대의 터미널을 한 대의 메인프레임에 연결해 여러 사용자끼리 나눠 사용하는 것도 가능했으며, 원격으로 타이핑을 한다는 의미에서 teletype이라는 명칭으로도 불렸다.

<img src="./image/pc.png" />

시대가 흘렀지만 teletype은 운영체제 안에 흔적이 남아있다. **`/dev`**(device, 장치) 폴더 안에는 tty(teletype)로 시작하는 여러 대의 소프트웨어 터미널이 있습니다. 물론 시리얼 포트를 통해 실제 하드웨어 터미널과 연결할 수 있는 ttyS도 함께 남아있다.

## 10. 리눅스 기반 운영체제 (2) 프로세스 관리와 시스템 콜

이제 터미널과 터미널 에뮬레이터의 차이를 배웠으니, 여러분의 터미널 에뮬레이터를 열고 몇 가지 명령어를 새로 배워보자. 

> 주피터 노트북이 돌아가고 있다면 새로운 탭 또는 창으로 새로운 터미널 세션을 시작하자.

우선 **`ps -e`**를 입력하여 현재 실행 중인 프로세스들을 확인한다. 윈도우 작업 관리자의 첫 화면과 같다고 생각하시면 됩니다.

> 👍️ **`ps`**: 현재 터미널과 관련된 프로세스 목록을 출력한다.<br>
> 👍️ **`ps -ef`**: 모든 터미널에서 현재 실행중인 프로세스 목록 및 부가정보를 출력한다.

```bash
june@juneoh-ubuntu:~$ ps -ef
june@juneoh-ubuntu:~$ ps -ef
  PID TTY          TIME CMD
    1 ?        00:00:01 systemd
    2 ?        00:00:00 kthreadd
    3 ?        00:00:00 rcu_gp
    4 ?        00:00:00 rcu_par_gp
    6 ?        00:00:00 kworker/0:0H-kb
    8 ?        00:00:00 mm_percpu_wq
    9 ?        00:00:00 ksoftirqd/0
   10 ?        00:00:00 rcu_sched
[..생략..]
 1980 tty2     00:00:00 /usr/bin/Xorg
 2027 pts/0    00:00:00 bash
 4486 ?        00:00:00 kworker/u2:2-ev
 4947 ?        00:00:00 kworker/u2:1-ev
 5147 pts/0    00:00:00 ps
```

**`PID`**는 process ID 번호이고, **`TTY`**는 해당 프로세스가 붙어있는 teletype(터미널)이다. 프로세스에서는 2번 터미널(**`tty2`**)에서 GUI 셸을 관리하는 프로그램인 Xorg가 돌아가고 있는 것을 확인할 수 있다. 또한 0번 가상 터미널(**`pts`**, pseudo teletype slave)에 우리가 방금 실행한 **`ps`** 명령어와, 해당 명령어를 실행한 셸 **`bash`**가 떠있다. **`TTY`** 칸에 **`?`**가 표시된 프로세스들은 터미널에 부착되지 않은 프로세스들이다.

자, 이제 한번 프로세스를 하나 죽여보자. GUI 셸의 핵심인 **`Xorg`**가 만만해 보인다. 조금 갑작스럽지만 컴퓨터는 원래 망가트리면서 배워야 제맛이다. 물론 실제로 망가뜨리진 않겠지만, 열려있던 창이 모두 종료될 테니 작업 중이던 내용을 모두 저장한 다음, 위에서 확인한 **`/usr/bin/Xorg`** 프로세스의 번호를 넣어 **`sudo kill -s KILL [프로세스번호]`**를 입력한다. 

**[프로세스번호] 부분에 여러분의 컴퓨터에서 실제로 출력한 숫자를 집어넣자. 여러분의 터미널에서 나오는 프로세스 번호는 GUI가 켜질 때마다 다르게 부여될 수 있으니 아래와 같이 꼭 직접 확인해야 한다.**

```bash
june@juneoh-ubuntu:~$  ps -ef | grep Xorg
root      1980  1538  5 19:57 tty2     00:11:31 /usr/lib/xorg/Xorg vt2 -displayfd 3 -auth /run/user/1000/gdm/Xauthority -background none -noreset -keeptty -verbose 3
june@juneoh-ubuntu:~$  sudo kill -s KILL 1980
```

> 👍️ **`grep`**: 입력으로 전달된 내용에서 특정 문자열을 포함한 라인만 선택해서 출력한다. 파이프(**`|`**)를 통해 다른 명령문 실행결과와 결합해서 사용할 때 특히 유용하다.<br>
> ❤️ **`sudo`**: 이어지는 명령을 최고 관리자 권한으로 실행한다.<br>
> 👍️ **`kill`**: PID에 해당하는 프로세스에 시그널을 보낸다.

```bash
june@juneoh-ubuntu:~$ sudo kill -s KILL 1980
```

Xorg가 죽으면서 화면이 잠시 꺼졌다가, 자동으로 재시작하면서 다시 로그인 화면으로 돌아올 것이다. (혹시 응답이 없다면 그냥 컴퓨터를 껐다 켜자.) 참고로 나중에 언젠가 컴퓨터가 돌아가는데 화면이 얼었을 때, 이런 식으로 3번 터미널의 CLI로 전환해서 GUI를 죽이고 다시 2번 터미널의 GUI로 돌아오는 방법으로 재부팅 없이 해결할 수 있다. 물론 GUI 상에서 돌아가고 있던 프로그램들의 정보는 날아갈 수 있으니 주의해야 한다.

그럼 Xorg 프로세스의 종료는 어떤 방식으로 이뤄진 걸까? 프로세스 관리는 운영체제의 고유 권한이기 때문에, 우리는 커널의 도움이 필요하다. 우리가 CLI 셸에 **`kill`** 명령어를 입력하는 순간, 셸의 프로세스는 커널에게 이 명령을 전달하고, 커널은 권한 등을 확인한 후에 이를 수행한다. 이렇게 프로세스가 커널에게 요청을 전달하는 기능을 **시스템 콜(system call)**이라고 한다. 즉, 시스템 콜은 운영체제의 커널이 프로그램에 제공하는 인터페이스이며, 프로세스가 직접 할 수 없는 프로세스 제어(**`kill`** 등), 파일 조작(폴더 생성 **`mkdir`**, 파일 읽기 **`read`** 등), 장치 및 자원 관리, 네트워크 통신 등을 가능케 한다. 반대로 보면, 각 프로세스는 커널을 통하지 않고서는 파일이나 장치, 또는 다른 프로세스에 간섭할 수 없다는 뜻이 된다.

참고로 **`kill`**은 그 자체로 프로세스를 죽이는 명령이 아니라, 프로세스에 신호를 보내는 명령어이다. 강제종료하는 **`KILL`** 외에도, 프로그램 실행 중에 **`Ctrl+C`**를 누르면 전달되는 **`INT`** (interrupt) 신호나, 프로세스를 종료하기 전에 해당 프로세스에게 해명할 기회를 주는 **`TERM`** (terminate) 신호도 있다.

## 11. 리눅스 기반 운영체제 (3) 프로그램과 환경변수

여태까지 셸에서 여러 명령어를 실행해 보았지만, 사실 우리가 실행했던 명령어들의 진짜 정체는 두 가지로 구분할 수 있다. 1) 실제 프로그램과 2) 셸에서 제공하는 셸 명령어이다. **`which`** 명령어를 사용해서 그 정체를 한번 밝혀보자.

> 👌️ **`which`**: 명령어의 전체 경로를 출력한다.

```bash
june@juneoh-ubuntu:~$ which ls
/bin/ls
```

```bash
june@juneoh-ubuntu:~$ which cd
```

**`ls`**는 **`/bin`** 디렉토리에 설치된 **`ls`**라는 프로그램이고, **`cd`**는 프로그램이 아니라 셸에서 제공하는 명령어였기에 아무 결과도 출력되지 않는다. 우리가 사용하는 셸은 **`bash`**였다. 셸도 프로그램일 테니 한번 경로를 확인해보자.

```bash
june@juneoh-ubuntu:~$ which bash
/bin/bash
```

즉, 우리가 셸에 명령어를 입력하면 셸은 일단 셸 내에서 정의된 명령인지 확인한다. 만약 셸에 정의된 명령 중 있다면, 그대로 명령을 실행한다. 셸에 정의된 명령이 아니라면, 설치된 프로그램 중에서 해당 이름을 가진 프로그램이 있는지 확인하여 실행한다. 그럼 해당 이름을 가진 프로그램이 실제로 설치되어 있는지는 어디서 확인할까? **`echo $PATH`** 라고 입력하여 한번 물어보자.

```bash
june@juneoh-ubuntu:~$ echo $PATH
/usr/local/sbin:/usr/local/bin:/usr/sbin:/usr/bin:/sbin:/bin:/usr/games:/usr/local/games
```

무언가 나왔다. 일단 우리가 입력한 명령을 먼저 이해해 보겠다. **`echo`**는 뒤에 입력된 내용을 출력하라는 명령어였는데, 뒤의 **`$PATH`**는 무엇일까?

셸에서 단어 앞에 **`$`**가 붙으면 **환경 변수(environment variable)** 를 의미한다. 파이썬의 변수처럼, 환경 변수는 운영체제가 프로세스 단위로 사용하는 변수이다. 즉, 우리가 **`echo $PATH`**는 **`PATH`**라는 이름의 환경 변수(**`$`**)의 값을 출력(**`echo`**)하라는 뜻이다.

**`PATH`** 환경 변수는 셸이 프로그램을 실행할 때 참조할 경로들을 나타낸다. 값을 보면 디렉토리 여러 개가 **`:`**로 구분되어 있는 것을 알 수 있는데, 만약 동일한 이름의 프로그램이 여러 디렉토리에 설치되어 있다면, **`PATH`** 내 순서 상 앞에 있는 디렉토리의 프로그램을 설치한다. 아나콘다(Anaconda)를 설치했다면, 우분투에 기본으로 설치된 파이썬 대신 아나콘다가 설치한 파이썬을 사용하도록 **`PATH`**의 값이 변경되어 있을 것이다. 마찬가지로 **`which python`**을 통해 확인해볼 수 있다.

참고로 **`PATH`** 환경 변수의 기본값은 사용자가 로그인할 때 운영체제가 자동으로 정의하지만, 사용자가 직접 수정할 수도 있다. 셸 명령어인 **`export`** 명령어를 통해 **`PATH`**의 앞에 **`/home/june`** 디렉토리를 추가해보자.

> 👌️ **`export`**: 셸 변수나 함수를 현재 환경으로 내보낸다.

```bash
june@juneoh-ubuntu:~$ echo $PATH
/usr/local/sbin:/usr/local/bin:/usr/sbin:/usr/bin:/sbin:/bin:/usr/games:/usr/local/games
june@juneoh-ubuntu:~$ export PATH=/home/june:$PATH
june@juneoh-ubuntu:~$ echo $PATH
/home/june:/usr/local/sbin:/usr/local/bin:/usr/sbin:/usr/bin:/sbin:/bin:/usr/games:/usr/local/games
```

또한 **`env`** 명령어로 현재 설정된 환경변수들을 모두 출력할 수 있다.

> 👌️ **`env`**: 명령어가 제공된다면 해당 명령어를 새로운 환경에서 실행한다. 명령어가 제공되지 않은 경우 현재 환경의 정보를 출력한다.

```bash
june@juneoh-ubuntu:~$ env
HOSTNAME=juneoh-ubuntu
USER=june
PWD=/
HOME=/home/june
MAIL=/var/mail/june
SHELL=/bin/bash
TERM=xterm
SHLVL=2
LOGNAME=june
PATH=/home/june:/usr/local/sbin:/usr/local/bin:/usr/sbin:/usr/bin:/sbin:/bin:/usr/games:/usr/local/games
_=/usr/bin/env
```

우리가 **`/bin`** 디렉토리 안의 프로그램 **`bash`**를 현재 셸(**`SHELL`**)로써 사용한다는 내용도 나와있다.

## 12. 리눅스 기반 운영체제 (4) 패키지 관리자****

그럼 리눅스 기반 운영체제, 특히 우분투에서 새로운 프로그램을 설치하거나 기존에 설치된 프로그램을 업데이트 하려면 어떻게 할까? 인터넷에서 설치 파일을 다운로드 받아 직접 실행할 수도 있지만, 우분투 운영체제는 공식적으로 APT(Advanced Packaging Tool)라는 **패키지 관리자(package manager)** 를 제공한다.

패키지 관리자란 공인 저장소에서 프로그램과 라이브러리(다른 프로그램들이 참조하여 사용할 수 있는 코드)를 패키지 단위로 설치하고 관리할 수 있도록 하는 프로그램이다. 내가 작성한 유용한 코드를 잘 패키징해 모두가 볼 수 있는 곳에 올려둔다면, 다른 사람들도 손쉽게 코드를 다운 받아 사용할 수 있을 것이다.

우분투에는 APT말고도 Snap이라는 비교적 새로운 프로그램도 있으며, CentOS의 YUM(Yellowdog Updater Modified), Red Hat 계열의 RPM(Red Hat Package Manager) 등 각 리눅스 계열 운영체제는 다양한 기본 패키지 관리자를 제공한다. 운영체제 상 패키지 관리자 외에도, 파이썬의 PIP(Python Package Installer) 등 프로그래밍 언어상 제공되는 경우도 있다.

이제 APT로 우분투에 **`htop`**이라는 프로그램을 설치해 보겠다. **`htop`**은 CPU 및 메모리 사용량을 시각화하고, 프로세스 별 CPU 사용량을 표시해 주는 프로그램이다.

> ❤️ **`apt-get`**: APT 패키지 관리자를 통해 공인 저장소에서 패키지를 가져다 설치하거나 제거한다.

우선 등록된 공인 저장소에 있는 패키지 목록을 최신 버전으로 가져온다.

```bash
june@juneoh-ubuntu:~$ sudo apt-get update
Get:1 http://security.ubuntu.com/ubuntu bionic-security InRelease [88.7 kB]
Hit:2 http://archive.ubuntu.com/ubuntu bionic InRelease
Get:3 http://archive.ubuntu.com/ubuntu bionic-updates InRelease [88.7 kB]
Get:4 http://archive.ubuntu.com/ubuntu bionic-backports InRelease [74.6 kB]
Fetched 252 kB in 2s (105 kB/s)
Reading package lists... Done
Building dependency tree
Reading state information... Done
All packages are up to date.
```

이제 설치를 진행한다. 패키지 목록을 읽고, 해당 패키지가 의존하는 다른 패키지들이 있는지 등을 자동으로 확인한 뒤, 설치를 진행한다.

```
june@juneoh-ubuntu:~$ sudo apt-get install htop
Reading package lists... Done
Building dependency tree
Reading state information... Done
Suggested packages:
  lsof strace
The following NEW packages will be installed:
  htop
0 upgraded, 1 newly installed, 0 to remove and 0 not upgraded.
Need to get 80.0 kB of archives.
After this operation, 221 kB of additional disk space will be used.
Get:1 http://archive.ubuntu.com/ubuntu bionic/main amd64 htop amd64 2.1.0-3 [80.0 kB]
Fetched 80.0 kB in 1s (57.0 kB/s)
debconf: delaying package configuration, since apt-utils is not installed
Selecting previously unselected package htop.
(Reading database ... 4046 files and directories currently installed.)
Preparing to unpack .../htop_2.1.0-3_amd64.deb ...
Unpacking htop (2.1.0-3) ...
Setting up htop (2.1.0-3) ...
```

**설치가 완료되면 바로 실행해 봅시다. `htop`을 종료하려면 `q`를 누르면 됩니다.**

```bash
june@juneoh-ubuntu:~$ htop
```

삭제는 설치만큼 쉽다. **`apt-get`** 뒤에 **`install`** 대신 **`remove`**를 넣어서 프로그램만을 제거하거나, **`purge`**를 넣어 해당 프로그램의 설정 파일까지 모두 삭제시킬 수 있다. 만약 중간에 확인(**`[Y/n]`**)을 물으면 **`y`**를 누르고 엔터를 쳐서 제거를 승인한다.

```bash
june@juneoh-ubuntu:~$ sudo apt-get remove htop
Reading package lists... Done
Building dependency tree
Reading state information... Done
The following packages will be REMOVED:
  htop
0 upgraded, 0 newly installed, 1 to remove and 0 not upgraded.
After this operation, 221 kB disk space will be freed.
Do you want to continue? [Y/n]
```

## 13. 리눅스 기반 운영체제 (5) 사용자와 권한

### **`sudo` 명령어**

---

여태까지 **`sudo`**라는 명령어를 계속 사용해왔는데, 이제 그 뜻을 한번 자세히 알아볼 때가 되었다.

하나의 메인프레임 컴퓨터에 여러 개의 터미널을 붙여 사용할 수 있게 했던 것처럼, 한 대의 컴퓨터를 여러 사람이 함께 사용하는 것은 운영체제의 초기 역사부터 일반적인 사용법이었다. 이 때문에 사용자(user)를 구분하고 권한을 나누는 일이 중요했으며, 그 개념은 한 사람이 한 대의 컴퓨터를 사용하는 개인용 컴퓨터(personal computer, PC) 시대의 운영체제에도 그대로 남았다. 특히 유닉스 계열 운영체제들은 개인용 컴퓨터보다 서버용 또는 기타 산업용으로 쓰이는 경우가 잦기에, 다른 운영체제보다 사용자와 권한의 개념을 훨씬 더 자주 마주치게 된다.

유닉스 계열 운영체제에서 모든 권한은 사용자를 기준으로 인증한다. 운영체제에서 어떤 작업을 실행하기 위해서는 어떤 사용자 이름으로 실행할지 미리 정해져 있어야 한다는 뜻이다.

**`ps -ef`** 명령어를 사용하면 프로세스마다 누구의 권한으로 실행되는지(**`UID`**, User ID) 확인할 수 있다. 프로세스는 현재 사용자가 가진 범위만큼의 권한을 가진다. 우리가 위에서 **`cd /root`**로 최고 관리자의 디렉토리에 접근하려고 했을 때처럼, 해당 사용자의 권한을 넘어서는 작업을 요청하면 운영체제의 커널이 이를 거부한다.

보통 우리가 사용하는 사용자 계정은 운영체제를 설치할 때 최초로 생성한 계정이다. 개인용 컴퓨터에서는 이 사용자가 해당 컴퓨터를 사용할 주된 사람일 것이다. 하지만 리눅스 기반 운영체제, 윈도우, 맥 OS 모두 공통적으로, 해당 사용자 계정에게 운영체제에 대한 모든 권한이 주어지지는 않는다. 보안 및 시스템 안정성을 위해 완전한 권한은 운영체제에 기본으로 설정된 **최고 관리자(superuser)** 계정만이 가지고 있으며, 윈도우에서는 이 계정을 Administrator이라고 부르고, 유닉스 계열 운영체제에서는 **`root`**라는 아이디를 가지고 있다.

우리가 직접 생성한 사용자 계정은 일반 사용자 계정이다. 이 때문에 운영체제 자체에 영향을 주는 일(커널 권한이 필요한 프로그램의 설치나 실행 등)은 최고 관리자 권한을 빌려야 한다. 윈도우에서 무언가를 설치할 때 가끔 보게 되는 "관리자 권한으로 실행" 경고창이나, 유닉스 계열 운영체제에서 사용하는 **`sudo`**(Switch User Do) 명령어가 바로 이 역할을 하며, 일반 사용자가 잠시 최고 관리자(또는 다른 사용자) 권한으로 행동할 수 있게 해준다. 매번 확인해야 하는 것이 번거로울 수 있지만, 한 번 더 확인함으로써 실수를 방지할 기회를 준다는 것 외에도 모든 권한이 아닌 일부 권한만 허용할 수 있는 등 다양한 보안상의 이점이 있다.

### **사용자 그룹**

---

사용자들이 여러 명 있을 때, 쉽게 묶어서 관리하는 개념이 바로 그룹(group)이다. 예를 들어 한 사용자가 **`sudo`** 명령어를 실행하기 위해서는, 동일한 이름(**`sudo`**)라는 이름의 그룹에 속해있어야 한다.

컴퓨터에 어떤 그룹들이 있고, 해당 그룹마다 어떤 사용자들이 들어있는지 한번 확인해보자. 유닉스 계열 운영체제에서는 설정 파일이나 장치 모두 파일로 접근 가능해야 한다는 철학이 있기 때문에, 그룹 설정도 텍스트 파일로 저장되어 있다.

> ❤️ **`cat`**: 하나 이상의 텍스트 파일을 순서대로 출력한다.

```bash
june@juneoh-ubuntu:/$ cat /etc/group
root:x:0:
daemon:x:1:
bin:x:2:
sys:x:3:
adm:x:4:
tty:x:5:
[..생략..]
sudo:x:27:june
[..생략..]
staff:x:50:
games:x:60:
users:x:100:
nogroup:x:65534:
june:x:1000:
```

<img src="./image/cat.png" />
<center>[[`cat` 명령어는 고양이가 아니라 concatenate(이어붙이다)의 약자입니다.]]</center>

**`sudo`** 그룹(27번)에 **`june`** 사용자가 들어있는 것을 알 수 있다. 특이한 점은 직접 만든 적이 없음에도 사용자명과 동일한 이름의 그룹(**`june`**, 1000번)이 등록되어 있다는 사실이다. 이는 모든 사용자는 최소한 하나 이상의 그룹에 등록되어 있어야 하기 때문에 사용자가 생성될 때 자동으로 동일한 이름을 가진 그룹이 생성되기 때문이다.

참고로 사용자들의 목록은 **`/etc/passwd`**에 기록되어 있는데, 원래 예전에는 해당 파일에서 사용자의 비밀번호까지 기록했기 때문이다. 이제 대부분의 리눅스 기반 운영체제에서 비밀번호는 최고 관리자만 접근 가능한 **`/etc/shadow`** 파일에 암호화되어 저장된다.

### **접근 권한**

---

그럼 누가 어떤 프로그램을 실행할 수 있는지는 어떻게 구분할까? 다시 한 번, 프로그램도 파일이기 때문에, 프로그램 실행 권한은 파일 접근 권한과 동일한 방법으로 구분한다.

프로그램과 디렉토리를 포함한 모든 "파일"에는 소유 사용자와 소유 그룹이 있다. 소유 사용자가 할 수 있는 일, 소유 그룹에 소속된 사용자들이 할 수 있는 일, 그리고 그 밖의 사용자들이 할 수 있는 일을 파일마다 설정할 수 있다. 이 설정값은 **`ls -lah`** 명령을 통해서도 확인할 수 있다.

```bash
june@juneoh-ubuntu:~$ ls -lah
total 64K
drwxr-xr-x 13 june june 4.0K Mar 17 01:20 .
drwxr-xr-x  6 root root 4.0K Mar 17 00:15 ..
-rw-r--r--  1 june june  220 Apr  4  2018 .bash_logout
-rw-r--r--  1 june june 3.7K Apr  4  2018 .bashrc
drwx------  2 june june 4.0K Mar 17 00:16 .cache
drwxrwxr-x  2 june june 4.0K Mar 17 01:17 Desktop
drwxrwxr-x  2 june june 4.0K Mar 17 01:20 Documents
drwxrwxr-x  2 june june 4.0K Mar 17 01:17 Downloads
drwx------  3 june june 4.0K Mar 17 00:16 .gnupg
drwxrwxr-x  2 june june 4.0K Mar 17 01:20 Music
drwxrwxr-x  2 june june 4.0K Mar 17 01:20 Pictures
-rw-r--r--  1 june june  807 Apr  4  2018 .profile
drwxrwxr-x  2 june june 4.0K Mar 17 01:20 Public
drwx------  2 june june 4.0K Mar 17 00:15 .ssh
drwxrwxr-x  2 june june 4.0K Mar 17 01:20 Templates
drwxrwxr-x  2 june june 4.0K Mar 17 01:20 Videos
```

'june'이 여러 번 나온다. 3번 열은 소유 사용자이고, 4번째 열은 소유 그룹이다. 위에 두 줄을 보면 현재 디렉토리(**`/home/june`**, 여기서는 **`.`**)는 사용자로서 **`june`**과 그룹으로서 **`june`**이 소유하고 있으며, 그 위의 디렉토리(**`/home`**, 여기서는 **`..`**)는 사용자와 그룹으로서의 **`root`**가 소유하고 있음을 알 수 있다. 한번 이걸 변경해보자. 현재 디렉토리의 소유 사용자를 **`root`**로, 소유 그룹은 **`june`**으로 변경했다가 복구해 보겠다.

> 👍️ **`chown`**: 대상 파일의 소유 사용자와 그룹을 변경한다.

```bash
june@juneoh-ubuntu:~$ sudo chown root:june .
june@juneoh-ubuntu:~$ ls -lah
total 24K
drwxr-xr-x 13 root june 4.0K May  1 04:07 .
drwxr-xr-x  6 root root 4.0K May  1 04:06 ..
[..생략..]
june@juneoh-ubuntu:~$ sudo chown june:june .
june@juneoh-ubuntu:~$ ls -lah
total 24K
drwxr-xr-x 13 june june 4.0K May  1 04:07 .
drwxr-xr-x  6 root root 4.0K May  1 04:06 ..
[..생략..]
```

참고로 위 명령어만으로는 디렉토리 내용물까지 소유자가 변경되지는 않는다. 디렉토리 내의 모든 파일들에 대해 적용하려면 **`-R`** 옵션을 사용해야한다.

소유자 외에도, 위 **`ls -lah`**의 첫 열을 보면 **`drwxr-xr-x`**라고 무언가 적혀있는데, 이 부분이 바로 권한을 표기하는 부분이다. 첫 글자인 **`d`**는 해당 파일이 디렉토리인지 또는 일반 파일인지 여부를 나타낸다. 그 뒤를 잇는 글자들은 각각 3 글자씩 *소유 사용자*, *소유 그룹*, *기타 사용자*에게 허용된 *읽기(read, r)*, *쓰기(write, w)*, *실행하기(execute, x)* 권한을 나타낸다.

예를 들어 **`drwxr-xr-x`**는 해당 파일이 디렉토리이고(**`d`**), 소유자는 읽고(**`r`**) 쓰고(**`w`**) 실행(**`x`**)하는 모든 동작이 가능하지만, 소유 그룹 소속 사용자와 기타 사용자들은 읽고 실행하기만 가능하다는 것을 표시한다.

권한은 숫자로도 표시할 수 있다. 기본적으로 세 가지 권한 **`rwx`**를 각각의 이진수 자리로 본다. 세 자리 이진수 숫자를 해석하면 사용자, 그룹, 기타 사용자에게 부여된 권한을 1자리 십진수 숫자로 나타낼 수 있다. 예를 들어 **`rwxr-xr-x`**를 이진수로 나타내면 **`111101101`**이고, 이는 구분 별로 다음과 같다.

* 사용자: 이진수 **`111`** = 십진수 **`7`**
* 그룹: 이진수 **`101`** = 십진수 **`5`**
* 기타 사용자: 이진수 **`101`** = 십진수 **`5`**
* 종합: **`755`**

참고로 실행하기(**`x`**)의 경우 프로그램이라면 실제 실행할 수 있다는 것을 나타내지만, 디렉토리의 경우 해당 폴더 안으로 **`cd`**로 이동할 수 있는지 여부를 나타낸다.

### **`chmod` 명령어**

---

**`chmod`** 명령어를 사용하면 권한을 변경할 수 있다. **`aiffel`** 폴더의 권한을 한번 변경해보자. 모두에게 모든 권한을 끄기 위해서 숫자 표기로 **`000`**을 입력한다. (**`000`** = **`000000000`** = **`---------`**)

> 👍️ **`chmod`**: 대상 파일의 권한을 변경한다.

```bash
june@juneoh-ubuntu:~$ chmod 000 aiffel
june@juneoh-ubuntu:~$ ls -lah
total 28K
drwxr-xr-x 3 june june 4.0K May  1 06:01 .
drwxr-xr-x 1 root root 4.0K May  1 04:06 ..
-rw------- 1 june june   15 May  1 04:06 .bash_history
-rw-r--r-- 1 june june  220 May  1 04:06 .bash_logout
-rw-r--r-- 1 june june 3.7K May  1 04:06 .bashrc
-rw-r--r-- 1 june june  807 May  1 04:06 .profile
-rw-r--r-- 1 june june    0 May  1 04:07 .sudo_as_admin_successful
d--------- 2 june june 4.0K May  1 06:01 aiffel
june@juneoh-ubuntu:~$ ls aiffel
ls: cannot open directory 'aiffel': Permission denied
```

이제 **`aiffel`** 폴더에 접근할 수 없으니 앞으로 공부를 계속할 수 없다.

__공부를 계속해야 한다면, 어떤 명령어를 수행해야 할까?__

* chmod 755 aiffel

### **기계어**
---
이렇게 제작된 CPU 칩이 처리할 수 있는 명령어는 **기계어**라고도 하며, 우리가 사용하는 프로그래밍 언어와는 많이 다르다. 기계어로 프로그램을 직접 짜기에는 워낙 복잡하고 비효율적이기에 C, C++, Python 등 인간이 이해할 수 있는 프로그래밍 언어들이 만들어졌다. 이렇게 인간이 이해할 수 있는 언어로 짠 프로그램을 CPU에서 실행할 수 있는 기계어로 번역하고 최적화하는 과정이 바로 **컴파일**이다.

### **명령어 집합**
---
회로 설계는 Intel, AMD 등 제조사마다 다르지만, 지원하는 연산 세트를 나타내는 **명령어 집합(instruction set)**은 데스크탑용 CPU 간 상당 부분 호환된다. 다만 스마트폰 등 모바일 기기에 들어가는 CPU는 이야기가 다른데, 대부분 ARM 사의 RISC(Reduced Instruction Set Computer)라는 구조를 가지고, 지원하는 명령어는 더 적지만 전력을 더 적게 소비하는 방향으로 설계된다.

> 흥미롭게도 ARM 회사는 더 이상 직접 칩을 만들지 않고, 설계만 만들어 라이센스를 판매한다. 이 설계를 원형으로 삼성전자(Exynos 등), 애플(A11 등), 퀄컴(SnapDragon 등) 등 회사가 자신들의 목적에 맞게 변경하여 각자 칩을 생산한다.

### 메모리와의 관계
---
CPU를 구분할 때에는 32bit와 64bit로 구분하기도 하는데, 이는 CPU가 한 번에 읽고 쓸 수 있는 데이터의 크기를 나타낸다. 이 크기가 중요한 이유 중 하나는 사용 가능한 **메모리 주소**의 범위에 영향을 주기 때문이다.

이론적으로 32bit 컴퓨터에서는 2의 32승인 약 42억 개 메모리 주소공간만 사용할 수 있다. 1개 주소공간이 1 Byte라면 4GB 크기의 메모리 용량만 표기할 수 있기 때문에, 그 이상의 RAM을 사용할 수 없다는 큰 제한사항이 있었다.

### CPU의 성능과 발전
---
CPU의 성능을 판단하는 주요 척도 중 하나는 초당 처리 속도의 기준인 클럭(clock)이다. 하지만 요새 나오는 CPU를 보면 클럭이 4GHz에서 더 이상 발전하기보다는, 동시에 더 많은 스레드를 처리할 수 있도록 코어 수를 늘리는 방향으로 제조되고 있다. 이는 처리 속도를 높이기 위해 소비 전력을 높였다가는 물리적으로 냉각시키기가 너무 어려워지고, 그렇다고 전압을 낮춰서 낮은 전력으로 동작시키기에는 전류가 밖으로 누설되고, 그러면 전력과 전압을 모두 적절하게 유지하면서 집적도를 높이기에는 공정이 너무 미세해지면서(1~5 나노미터) 수율도 떨어지고, 현재의 소재들로는 양자 터널링 현상으로 인해 전자가 닫혀있는 게이트를 그냥 뛰어넘어버리게 되는 복잡한 물리적인/경제적인 이유 때문이다.

### **다른 연산장치**

이러한 한계로 인해 최근에는 CPU 외에도 특정 작업에 특화된 연산장치들이 떠오르고 있다.

그래픽 연산을 위한 **GPU(Graphical Processing Unit)**이 그 중 하나이다. CPU보다 이해할 수 있는 명령어는 훨씬 적지만, 코어 수를 훨씬 늘리는 방법으로 다차원 행렬 연산에 특화했다. 비유하자면 8 코어 CPU가 박사학위를 소지한 수학자 8명이라면, GPU는 (사칙연산밖에 못하지만 계산을 틀리지는 않는) 초등학생 3,000명을 모아놓은 것이다. 복잡한 수식을 처리한다면 박사들이 유리하겠지만, 초등학교 수학 전집 100권을 빠르게 처리하기에는 초등학생들이 더 유리할 것이다.

딥러닝도 그래픽 연산과 유사하게 다차원 행렬을 계산하기 때문에 GPU를 많이 사용하지만, 최근에는 구글에서 딥러닝 전용으로 만든 **TPU(Tensor Processing Unit)** 처럼 전용 칩이 개발되고 있다.

이 외에 작업에 맞게 직접 설계하여 생산하는 **ASIC(Application-Specific Integrated Circuit, 어플리케이션 특화 집적회로)** 도 있으며, 그때그때 회로를 직접 프로그래밍해서 사용할 수 있는 **FPGA(Field-Programmable Gate Array)** 칩도 다양한 산업 분야에서 활용되고 있다.

## 16. 기타 (3) 클라우드와 가상화

### **클라우드 컴퓨팅**

---

닷컴버블 초기, 미국의 전자상거래 업체 아마존(Amazon)은 주요 연휴마다 고객들이 몰려 서버가 다운되는 일을 겪었다. 그렇다고 연휴 기간의 최대 수요에 맞춰 평상시에도 많은 서버들을 유지하고 있자니 비용이 만만치 않다는 문제가 있었다. 이 문제가 자신들에게만 해당되지 않는다는 점을 깨달은 아마존은, **아마존 웹 서비스(Amazon Web Services)** 라는 자회사를 설립하여, 수요에 따라 그때그때 서버를 빌렸다가 반납하는 서버 임대업을 시작한다.

하지만 이 서버 임대업은 데이터센터에 위치한 물리적인 서버를 임대하는 것이 아니었다. 마치 런던에서는 치솟은 집값 때문에 방을 쪼개어 각각 임대하는 것처럼, 아마존은 한 대의 물리적인 서버를 여러 개의 가상 서버로 쪼개어 판매하기 시작했다. 클라우드 컴퓨팅이라는 말은 그 이전부터 존재했지만, 이때부터 본격적인 상업적 클라우드 컴퓨팅 시대가 시작된다.

### **가상화**

---

**가상화(virtualization)** 는 이처럼 하나의 컴퓨터를 쪼개어 여러 대의 컴퓨터가 있는 것처럼 사용하게 하는 기술이다. 가상화에서 실제 물리적인 서버를 **호스트(host)**, 그 안에서 돌아가는 가상 서버를 **게스트(guest)** 라고 한다.

가상화에는 여러 가지 종류가 있다. **가상머신(Virtual Machine)** 가상화의 경우 게스트 자체에 운영체제가 돌아가지만, **컨테이너(Container) 가상화** 의 경우, 마치 화장실과 주방을 공유하는 것처럼, 게스트들이 호스트의 운영체제 커널을 공유한다. 물론 커널의 기능만 공유할 뿐, 프로세스나 자원 등은 서로 철저히 격리하여 서로 간섭할 수 없도록 하는 것이 핵심기술이다.

**도커(Docker)** 는 이런 컨테이너 가상화를 손쉽게 제공하는 프로그램이다. 한 곳에서 실행 환경(우분투 버전, 라이브러리, 필요한 프로그램들)을 컨테이너로 구성해놓은 다음에 그걸 그대로 클라우드로 가져가서 실행하거나, 굳이 클라우드가 아니어도 내 컴퓨터에서 여러 개의 독립된 작업 환경을 구분하는 데에 사용할 수 있다.

그렇다고 가상화가 무조건 클라우드 컴퓨팅과 함께하는 개념은 아니다. 내 컴퓨터에서 파이썬 버전을 여러 가지 사용하고 각각 패키지를 구분하여 관리할 수 있게 해주는 아나콘다(Anaconda)나 VirtualEnv는 클라우드 컴퓨팅과 크게 관계가 없지만, 가상화의 한 종류로 볼 수 있다.

## 18. 셸 심화 (2) 셸 스크립트

명령어가 길거나, 명령어 여러 개를 이어서 실행해야 할 때, 명령어들을 파일로 묶어 실행할 수 있다. 한 줄에 한 명령어씩 적어주면 되고, **`#`**로 시작하는 주석을 넣을 수도 있다. 텍스트 편집기를 열어 **`test.sh`**라는 파일을 만들고, 다음 내용을 입력해보자.

```python
#!/bin/bash
echo "Hello"
# 주석
echo "Bye"
```

맨 위의 **`#!/bin/bash`**는 이 스크립트 파일을 실행할 때 어떤 프로그램을 사용해야 할지 알려준다. 또한 실제로 이 파일을 실행하기 위해 해당 파일에 **`chmod`**로 실행 권한을 추가해야 한다.

```python
june@juneoh-ubuntu:~$ chmod +x test.sh
june@juneoh-ubuntu:~$ ./test.sh
Hello
Bye
```

참고로 파이썬도 위의 규칙을 따라 스크립트 파일로 실행할 수 있다.

```python
#!/usr/bin/python
print("Hello")
```

셸에도 파이썬처럼 조건문과 연산자들이 있습니다. **`||`**를 써서 이전 명령이 false를 반환한 경우에만 다음 명령을 실행하거나, **`&&`**로 이전 명령이 true를 반환한 경우에만 다음 명령을 실행할 수 있다.

```bash
june@juneoh-ubuntu:~$ false || echo "false"
false
june@juneoh-ubuntu:~$ false && echo "false"
june@juneoh-ubuntu:~$ true && echo "true"
true
june@juneoh-ubuntu:~$ true || echo "true"
```

**`>`**를 통해 명령어의 결과를 파일에 저장하는 것도 가능한다. **`>>`**를 쓰면 명령어의 결과가 기존 파일 뒤에 추가된다.

```bash
june@juneoh-ubuntu:~$ echo hello > test.txt
june@juneoh-ubuntu:~$ cat test.txt
hello
june@juneoh-ubuntu:~$ echo hello again >> test.txt
june@juneoh-ubuntu:~$ cat test.txt
hello
hello again
```

또는 **`|`**를 통해 앞의 명령어의 결과를 입력으로 받아 이어서 처리할 수 있다. 예를 들어 아래와 같은 내용의 텍스트 파일(**`fruits.txt`**)이 있다면,

```txt
Orange
Apple
Grape
```

다음 명령을 통해 순서를 정렬할 수 있다.

> 👍️ **`sort`**: 입력된 텍스트를 줄 단위로 정렬한다.

```bash
june@juneoh-ubuntu:~$ cat fruits.txt
Orange
Apple
Grape
june@juneoh-ubuntu:~$ cat fruits.txt | sort
Apple
Grape
Orange
```

    * ❤️**`echo`**: 뒤이어 입력한 내용을 출력한다.

    * 👍**`tail`**: 텍스트 파일의 마지막 몇 줄만 출력한다.

    * 👍**`uniq`**: 텍스트 중 중복 값들을 제거한다.
    * 👍**`grep`**: 정규식을 활용해 텍스트를 검색한다.
    * 👍**`wc`**: 텍스트를 단어, 줄, 문자 등 단위로 센다.

    * 👌**`awk`**: 텍스트를 스캔하고 변형한다.

    * ❤️**`ls`**: 현재 디렉토리 내의 파일 및 폴더들을 출력한다.

    * ❤️**`cp`**: 파일을 복사한다.

    * ❤️**`ssh`**: 원격으로 셸에 접속한다.
    * 👍**`tmux`**: 여러 개의 셸 세션을 관리하고, 터미널 접속이 끊어지더라도 셸 세션을 유지한다.
    * 👍**`watch`**: 이어지는 명령을 주기적으로 실행하여 표시한다.
    * 👍**`curl`**: 주어진 URL과 통신한다.
    * 👍**`wget`**: 네트워크를 통해 파일을 다운로드한다.
    * 👍**`scp`**: SSH를 통해 원격으로 파일을 전송한다.
    * 👌**`ssh-keygen`**: SSH 공개키 비밀키 쌍을 생성한다.
    * 👌**`ssh-copy`**: SSH 공개키를 해당 서버에 등록한다.
    * 👌**`rsync`**: 원격으로 파일을 동기화한다.
    * 👌**`xarg`**: 명령어를 구성하여 실행한다. 이전 명령에서 인자를 넘겨받아 실행하거나, 다중 프로세스를 사용하여 명령을 병렬로 실행하는 등의 작업이 가능하다.
    * 👌**`cron`**: 주기적으로 실행될 명령을 등록한다.
    * 👌**`htop`**: 프로세스 목록 및 자원 사용량을 시각적으로 표시한다.

2. titanic.csv에서 남성과 여성이 몇 명인지 세어보자. 우선 **`tail`** 을 통해 첫 줄을 제외한 나머지 줄을 출력하고, **`cut`**으로 성별이 들어있는 열만 추출한 다음, ④로 뭔가 쓱싹 한 다음에 **`uniq`**로 각각 세어보자.

**[titanic.csv 다운로드 받기](https://www.kaggle.com/c/titanic/data?select=test.csv)**

```bash
june@juneoh-ubuntu:~$ tail -n ① titanic.csv | cut -d② -f③ | ④ | uniq ⑤
314 female
577 male
```

```bash
june@juneoh-ubuntu:~$ cat titanic.csv | grep '①' | wc ②
4
```

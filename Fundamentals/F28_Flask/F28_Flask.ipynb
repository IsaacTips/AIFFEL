### 전처리 웹 사이트를 제작해보자
---
네이버, 구글, 잡코리아 사이트와 같은 웹 사이트들을 거의 매일 이용한다. 이런 웹 사이트들은 인터넷 익스플로러 혹은 크롬과 같은 웹 브라우저를 이용해 실행되는 프로그램이라고 생각하면 된다. 비유를 해보자면 __웹 브라우저__는 윈도우(windows)이고 웹 사이트는 이 윈도우에서 실행되는 __엑셀(excel)__ 같은 프로그램이다.

<img src="./image/front_back.png" alt="front_back" />

웹 사이트는 크게 두 가지로 구성된다. 프론트 엔드(front-end)와 백 엔드(back-end)이다. 프론트 엔드는 우리가 보고 있는 웹 사이트의 화면을 말한다. 백 엔드는 화면에서는 보이지 않지만, 웹 사이트를 화면에 보여주기 위해 데이터베이스를 활용해 실질적인 정보의 가공/처리를 수행하는 구성 요소라고 이해하면 쉽다.

친구와 통화할 때를 생각해보자. 친구에게 전화를 걸기 위해 전화번호를 입력하고 통화를 하며 듣고 말할 때 사용하는 단말기의 기능은 프론트 엔드에 해당한다. 백 엔드는 전화를 할 수 있도록 깔아 놓은 통신 기지국들, 그리고 그 사이에 보이지 않게 동작하는 통신사의 수많은 서비스 시스템에 해당한다. 우리는 전화기에 단순히 전화번호 몇자리만 입력했을 뿐인데, 내 친구와 가장 가까운 기지국을 찾아서 정확히 내 친구의 단말로 통화를 연결해 주고, 전화 사용량만큼 과금기록을 남겨두는 등의 보이지 않는 작업들이 어디선가 진행되는 것이다. 백 엔드(통신 서비스 시스템)가 있어야 프론트 엔드(통화)도 가능할 것이다.

## 4. 플라스크(Flask) 시작하기
__플라스크__(flask)는 __마이크로 웹 프레임워크__이다.

* __마이크로(Micro)__ : 플라스크의 특징인 웹서비스를 구성하는 최소한의 기능을 담아 단순하지만, 확장이 가능한 설계를 의미한다.

* __웹(Web)__ : 인터넷 브라우저를 통해서 보고 있는 공간을 의미한다.

* __프레임워크(Framework)__ : 어떠한 목적을 달성하기 위해 복잡하게 얽혀있는 문제를 해결하기 위한 구조며, 소프트웨어 개발에 있어 하나의 뼈대 역할을 한다. 프레임워크 안에는 기본적으로 필요한 기능들이 미리 구축되어 있어서 추가하고자 하는 기능들만 선별적으로 개발하면 된다. 그리고 이때, 프레임워크에서 정의한 규칙을 준수해서 개발해야 한다. 즉, 프레임워크는 사람에게 있어서 뼈와 같으며 이 뼈 위에 장기와 같은 기능들을 하나씩 추가하는 것이다.

#### 코드 설명

><em><b>데코레이터(decorator)</b>
>
>데코레이터는 함수를 반환하는 함수이다. 데코레이터 밑에 있는 함수가 실행되기 전에 먼저 데코레이터가 실행된다고 생각하면 된다.</em>

아래 코드를 통해서 데코레이터를 이해해보자

````python
def main_function():
     print ("test function")
````

위의 코드는 "test function" 이라는 텍스트를 화면에 표시해주는 함수이다. "test function"이 화면에 표시되는 데 걸리는 시간을 알고 싶다면 어떻게 해야 할까? 아래와 같이 텍스트가 표시되기 전과 후의 시간을 알면 된다.

````python
import datetime

# 코드 선언부
def main_function():     
     print ("test function")

# 코드 실행부   
print (datetime.datetime.now()) #- 텍스트 표시 전 시간
main_function()
print (datetime.datetime.now()) #- 텍스트 표시 후 시간
````

실행해보면 "test function"과 그 앞뒤로 두 개의 시간대가 나오게 된다. 그럼, 함수의 실행 시간을 측정하려면 매번 이렇게 작성해야 할까? 파이썬은 데코레이터라는 기능을 이용해서 효과적으로 사용할 수 있는 방법을 제시한다.

데코레이터를 만들어서 사용하는 방법이다. `datetime_decorator`라는 데코레이터를 만들어 보겠다.

````python
# 코드 선언부
def datetime_decorator(func):
        def decorated():
                print(datetime.datetime.now())
                func()
                print(datetime.datetime.now())
        return decorated

@datetime_decorator
def main_function():
        print ("test function")


# 코드 실행부  
main_function()
````

데코레이터를 만드는 방식은 함수를 만드는 방법과 같다. 함수 `datetime_decorator(func)`를 정의하고 아래에 기능하는 코드를 작성한다. 함수 안에 함수가 있는 구조로 만들어서 `datetime_decorator(func)` 함수가 내부 함수를 반환하는 구조로 하면 자동으로 데코레이터에 등록이 된다. 함수의 이름과 내부에 있는 함수명은 자유롭게 쓸 수 있다.

## 5. 플라스크(Flask) 프로젝트
플라스크에는 __프로젝트(project)__라는 개념이 있다. 하나의 웹 사이트는 하나의 프로젝트이다. 아무 웹 사이트를 검색해 들어가 보면 게시판, 공지사항 등 여러 페이지로 이뤄져 있는 것을 볼 수 있다. 여기서 웹 페이지는 __앱(app)__이라고 볼 수 있다. 그래서 하나의 프로젝트 안에는 여러 개의 앱이 존재할 수 있다. 우리가 만들 '이미지 전처리 페이지'와 'SQL 전처리 페이지 '는 각각 다른 앱(app)으로 만들 것이다.

### GET & POST
---
GET과 POST는 웹에서 데이터를 주고받을 때 쓰이는 통신 규약이다.

클라이언트에서 HTTP로 백 엔드와 통신하는 방법 중 대표적인 방식은 GET과 POST 방식이다. GET은 주로 링크를 클릭할 때 사용하고 POST는 데이터가 있는 게시물을 올릴 때, 즉 이미지나 대량의 텍스트 데이터 같은 데이터를 담아 보낼 때 주로 사용한다. 아래 사진은 GET 방식이다.

<img src="./image/get_post.png" alt="get_post" />

GET 방식의 예시를 들자면, 아래와 같이 바로 `app.route('/index')` 링크에 들어가는 경우가 있다.

```python
@app.route('/index')
def about():
    return 'About project'
```

GET 이외에 POST도 같이 받으려면 아래와 같이 `@app.route('/calculate', methods=['POST', 'GET'])` 데코레이터를 작성해야 한다. 아래 코드는 실행은 안 되지만 하나의 예시로 보자. 클라이언트에서 서버로 통신을 요청하면 서버에서는 요청 방법에 관해 확인을 한다.

`request.method`를 이용해서 POST로 왔으면 `if`문 이하가 실행되고, GET으로 왔으면 `elif`문 이하가 실행될 것이다.

```python

    if request.method == 'POST':
        print("0 = ", request.form.get('pre_toggle_0'))
        print("1 = ", request.form.get('pre_toggle_1'))
        print("2 = ", request.form.get('pre_toggle_2'))
    return render_template('index.html')

@app.route('/get_selected_table', methods=["POST"])
def selected_table():
    text = request.form.get('table_name')
    print(text)
    return render_template('index.html')

@app.route('/get_selected_table2', methods=["POST"])
def select_table2():
    text = request.form.get('textbox')

    return render_template('index.html', label=text)

@app.route('/upload_image', methods=['POST'])
def upload_image_file():
    if request.method == 'POST':
        file = request.files['uploaded_image']
        if not file: return render_template('index.html', label="No Files")
        label = file

        return render_template('index.html', label=label)
```

## 7. UI 코드 (1) HTML 기초

### 태그(tag)
---
`<a>`와 같이 홑화살괄호로 감싸진 것을 __태그__라고 부른다. 즉, `<`와 `>`가 앞뒤로 있는 명령어다.

태그에는 시작 태그와 종료 태그가 있다.

* 시작 태그 : `<a>`
* 종료 태그 : `</a>`

말 그대로 시작 태그는 시작을 알리고, 종료 태그는 종료를 알리게 된다.

또한, 시작 태그는 __속성__(attribute)과 __값__(value)을 가질 수 있다.

(예) `<a href="http://naver.com">`

`href`는 __속성__이라고 하며 `http://naver.com`을 __값__이라고 한다. `<a>`는 문서를 링크시키기 위해 사용하는 태그인데 여기에 `href`라는 속성에 `http://naver.com` 값을 넣어 문서를 링크시키는 목적지를 적용한다.

### 엘리먼트(element)

## 8. UI 코드 (2) 버튼과 텍스트박스

### 버튼
---
버튼은 다음 페이지로 넘어가거나, 데이터를 선택하면 다음 작업을 진행하라고 명령할 때 쓰인다.

<img src="./image/input.png" alt="input" />

테이블 명을 입력하고 선택 버튼을 HTML로 구현한 코드이다.

```html
<form action="/get_selected_table" method="POST">
    <input type="text" name="table_name" placeholder="테이블 명" required="required" />
    <button type="submit">선택</button>
</form>
```

#### HTML 코드 설명
버튼을 눌렀을 때 서버쪽에 무언가 요청이 전달되어야한다. 서버쪽에 전달할 요청을 생성하는 태그로 `<form>`이 사용된다.

`<form action="/get_selected_table" method="POST">`

* `action="/get_selected_table"` : `app.py`에 `get_selected_table`이라고 라우팅 된 함수를 실행한다.
* `method="POST"` : 데이터 전달 방식은 POST를 이용한다.

`<input type="text" name="table_name" placeholder="테이블 명" required="required" />`

* `input` : 입력 UI 컴포넌트를 의미헌다.
* `type="text"` : "text"를 입력하는 UI 컴포넌트 즉 텍스트 박스를 의미한다.
* `name="table_name"` : 파이썬에서 HTML의 UI 컴포넌트에 입력된 데이터를 가져올 때 사용하는 이름을 지정한다.
* `placeholder="테이블 명"` : 입력 칸에 도움을 주는 텍스트를 넣는다.
* `required="required"` : 같은 form안의 버튼을 눌렀을 때 빈칸이면 무조건 입력을 하도록 한다. 이때, 다른 form 안에 있는 것은 신경쓰지 않는다.

`<button type="submit">선택</button>`

* `button type="submit"` : form 내부에서 일어난 내용의 전송 기능을 담당한다.

`</form>`

이미지 전처리 부분의 UI를 먼저 설명하느라 이미지 파일을 넣고 파일을 선택하는 파트가 조금 늦게 다뤄지게 되었다.

아래는 `'이미지 업로드 하기'`에 해당하는 HTML 코드이다.

```html

        <span>
            {{ label }}
        </span>
    {% endif %}

```

#### HTML 코드 설명

`<form action="/upload_image" method="POST" enctype="multipart/form-data">`
* `action="/upload_image"` : `app.py`에 `upload_image`라고 라우팅 된 함수를 실행한다.
* `method="POST"` : 데이터 전달 방식은 POST를 이용한다.
* `enctype=multipart/form-data` : 이 속성은 파일의 데이터를 전송하는 파일이다. 지금까지 UI 컴포넌트들의 변경 값만 전송했기 때문에 데이터 전송에 필요한 이 속성이 없었다.

`<input type="file" value="Upload" name="image">`
* `type="file"` : 입력 타입을 '파일'로 한다. `[file](https://developer.mozilla.org/en-US/docs/Web/HTML/Element/Input)`을 사용하면 브라우저는 사용자가 업로드 할 파일을 선택할 수 있는 필드를 제공한다.
* `name="uploaded_image"` : 파이썬에서 HTML의 UI 컴포넌트에 입력된 데이터를 가져올 때 사용하는 이름을 지정한다.

`{% if label %}`
* 플라스크는 JInja2라는 문법을 지원한다. 이 문법은 HTML에서 for, if 와 같은 문법을 쓸 수 있도록 해준다. `{% if label %}` 이 코드가 있는 곳도 HTML 코드 안에 있다. 즉, label에 어떤 값이 들어오면 아래의 코드를 실행한다는 의미이다.

`<span> {{ label }} </span>`
* label 값을 가져와서 span 안에 넣는다. `<span> </span>` 안에 있기때문에 label의 텍스트가 표시된다. 예를 들어, 파이썬 코드의 return 변수 label에 `image1.jpg` 이라는 text가 들어왔다면 화면에 `image1.jpg`가 표시된다.

`{% endif %}`
* if문의 종료를 의미한다.


* `'/upload_image'` : 바로 위 HTML 코드에서 `<form action="/upload_image" method="POST">` 부분이 있다. 여기서 `action="/upload_image"`를 라우팅하고 있는 파이썬 함수가 여기라는 것을 `@app.route('/upload_image', methods=["POST"])` 가 말하고 있는 것이다.
* `methods=["POST"]` : 데이터 전달 방식은 POST를 이용한다.

`if request.method == 'POST':`
* html에서 POST 전송이 오면 if문 아래의 코드를 실행한다.

`file = request.files['uploaded_image']`
* HTML에서 `uploaded_image`이라는 name을 가진 UI 컴포넌트에 있는 파일을 가져온다.

`if not file: return render_template('index.html', label="No Files")`
* 만약 파일이 선택되지 않았다면 'No Files'라는 텍스트를 return 값으로 label에 넣는다. 위의 HTML 코드에서 `{{ label }}` 에 해당하는 코드에 텍스트가 들어가게 된다.

`return render_template('index.html', label=file)`
* 파일이 선택되었다면 return 값에 label 변수에 file을 넣어서 위 HTML 코드에서 `{{ label }}` 에 해당하는 코드에 자동으로 file에 대한 정보 텍스트가 들어가게 된다.

```python
from PIL import Image

이미지 처리 함수

        return image.resize((int(width), int(height)))

    return image.transpose(Image.ROTATE_180)

def image_change_bw(image):
    return image.convert('L')

플라스크

@app.route("/index")

    return render_template('image.html')

def preprocessing():
    if request.method == 'POST':
        file = request.files['uploaded_image']
        if not file: return render_template('index.html', label="No Files")

        img = Image.open(file)

        is_rotate_180 = request.form.get('pre_toggle_0')
        is_change_bw = request.form.get('pre_toggle_1')
        is_change_size = request.form.get('pre_toggle_2')

        if is_rotate_180 == 'on':
            img = image_rotate(img)

        if is_change_bw == 'on':
            img = image_change_bw(img)

        if is_change_size == 'on':
            img = image_resize(img, request.form.get('changed_width'), request.form.get('changed_height'))

        img.save('result_image.png')

        src_dir = os.path.dirname(os.path.abspath(__file__))
        image_path = os.path.join(src_dir, 'result_image.png')

        # 결과 리턴
        return render_template('image.html', label=image_path)
```

`sql_command()` 함수는 SQL을 이용해서 DB에 명령을 내리기 위한 함수이다.

```python
def sql_command(conn, command):
```    

`try, except`
* try-except 문으로 `try` 안의 코드가 정상적으로 작동하면 `try`만 실행된다. 만약 정상적으로 작동 안 하고 에러가 난다면 `except` 아래의 코드가 실행된다.

`command = command.lower()`
* 명령어를 소문자로 바꾼다. `command` 뒤에 붙은 `lower()` 함수가 문자열 값을 모두 소문자로 바꿔준다.

`if "select" in command:`
* SQL문 안에 "select"가 들어 있으면 아래의 코드를 실행한다.

```python
if "select" in command:

    command_split = command.split(" ")
    select_command = "SELECT * FROM " + command_split[command_split.index("from")+1]
    df = pd.read_sql(select_command, conn, index_col=None)
    html = df.to_html()

    conn.close()

    return html
```

`command_split = command.split(" ")`
* SQL문을 `" "`(빈칸)을 기준으로 나눠서 리스트에 넣는다. 예를 들어, SELECT * FROM table 이면 ['SELECT', '*', 'FROM', 'table']로 나뉜다.

`select_command = "SELECT * FROM " + command_split[command_split.index("from")+1]`
* select_command에 "SELECT * FROM "과 `command_split[command_split.index("from")+1]` 에서 나온 글씨를 합친다.
* `command_split[command_split.index("from")+1]` : 위에서 SQL을 빈칸으로 나눈 리스트(예. ['SELECT', '*', 'FROM', 'table'] )에서 from이 위치한 곳의 index 값에 1을 더한 위치에 있는 단어를 가져온다. 즉 from은 2번 index에 있기 때문에 2+1 = 3 의 index 위치에 있는 단어인 table을 가져온다. 그럼 위에 select_command에는 "SELECT * FROM " + "table" 이 되어 결과로 "SELECT * FROM table" 이 된다.

`df = pd.read_sql(select_command, conn, index_col=None)`
* 판다스 모듈에 있는 `read_sql` 함수를 사용한다. 이 함수는 `conn`에 있는 DB 연결을 이용해서 `select_command`에 있는 SQL문을 DB에 넣는다. DB에서 나온 결과를 `df`에 넣는다. 결과는 dataframe 형식으로 나오고, 형식은 데이터 분석 쪽 노드에서 배우게 된다.

`html = df.to_html()`
* `to_html()`을 이용해서 dataframe을 HTML 형식으로 바꿔준다. dataframe은 표 형식으로 되어 있다.

`return html`
* 표 모양의 HTML 코드가 반환된다.

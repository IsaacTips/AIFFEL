## 2. 함수와 변수 (1) 함수 불러보기

### 함수 호출하기
---
수학에서의 함수란 (아주 단순화하면) 입력값(input)에 대해 출력값(output)을 가지는 관계를 가리킨다. f: x → y 또는 f(x) = y로 표기하는데, 여기서 f는 함수, x는 입력값, y는 출력값을 가리킨다.

그렇다면 프로그래밍에서의 함수는 어떨까? 여기서 함수의 의미는 조금 달라진다. 함수는 불려진 시점에 특정한 작업을 수행하며, 입력값과 출력값은 있을 수도 없을 수도 있다.

함수를 하나 불러보자. `print()`라는 함수는 파이썬에 의해 기본적으로 제공되는 함수로, 화면에 문자열을 표시해주는 역할을 한다.

````python
print()
````

위 `[Output]`을 확인해보면 아무것도 보이지 않는다. 왜일까? 아무런 입력값이 없어서 그렇다. 어떤 함수들은 입력값이 필수라서 입력값이 없을 때 오류를 뱉는 함수도 있다.

그렇다면 숫자 `1`을 입력값으로 넣어 다시 실행해보자.

````python
print(1)
````

수학에서 함수를 'y=f(x)'로 표기하는 것은 우리에게 너무나 익숙한 사실이다. 이 함수에서 f와 x가 각각 함수의 이름과 입력값을 의미하듯이, `print`와 괄호 속 `1`도 동일한 의미를 지닌다. 프로그래밍 분야에서는 입력값을 주로 __인자(argument)__ 라고 부르는데 함께 기억해두면 좋을 것이다.

`print()` 함수의 괄호 안에서 숫자끼리 덧셈을 하면 어떻게 될까?

````python
print(1 + 1)
````

수학의 계산식과 같이 괄호 안의 내용이 먼저 계산되는 것을 볼 수 있다. 숫자 대신 글자를 표시하고 싶다면, 프로그램의 코드와 구분해주기 위해 따옴표 `'` 또는 쌍따옴표 `"`로 내용을 감싸줘야 한다. 이렇게 감싼 글자들의 한 묶음을 하나의 __문자열__ 이라고 한다.

문자열 `'Hello'`를 출력해보자

````python
print('Hello')
````

위에서 `print(1 + 1)`을 통해 `2`가 제대로 출력된 것을 확인했었다.

문자열끼리 더하면 어떻게 될까?

````python
print('Hello ' + 'world!')
````

결과에서 추축할 수 있듯 파이썬에서는 문자열끼리 더하면 각각의 내용(문자열)을 이어준다.

그렇다면 빼는 것도 가능할까? 한번 시도해보자.

````python
print('Hello' - 'o')
````

## 3. 함수와 변수 (2) 변수 사용하기
f(x) = y에서 f()가 '함수'라면, 입력값과 출력값인 x와 y는 '변수'에 해당된다.

`print()`라는 함수에 `greeting`이라는 변수를 입력값으로 주는 예시를 한번 살펴보자.

````python
greeting = '안녕하세요?' 
#- greeting이라는 변수에 '안녕하세요?'라는 문자열을 저장합니다.
print(greeting) 
#- greeting을 출력합니다.
````

사전에서 '변수'를 검색해보면, 어떤 관계나 범위 안에서 여러 가지 값으로 __변할 수 있는 수__ 라고 정의되어 있다. 이처럼 변수는 스스로 값을 갖기보단 다른 값을 가리키는 존재이다.

위 코드의 첫번 째 줄에서 `greeting`이라는 변수가 `안녕하세요?`라는 문자열을 가리키도록 정의했고, 세번 째 줄에서는 `print()` 함수에 `greeting` 변수를 입력으로 주어 `greeting` 변수가 가리키는 값을 화면에 표시하도록 했다. 즉, 변수를 정의하기 위해서는 `[변수명] = [변수값]`과 같은 형식을 취하면 되며, 이후에는 변수명으로 해당 변수값을 참조할 수 있다.

여기서 또 배울 수 있는 점이 있다. 코드가 한 줄 한 줄 순서대로 처리되기 때문에 변수를 정의(define)하기 전에 먼저 사용하려고 하면 오류가 나게 된다.

````python
print(something)
````

오류 메시지를 읽어보면 `something`이라는 변수명이 정의되지 않아서(not defined) 발생된 오류임을 확인할 수 있다.

또 다른 경우를 살펴보자. 방금 위에서 변수란 어떤 값을 가리키는 존재라고 언급했다. 같은 변수명을 재사용하게 될 경우 어떤 값을 가리키는지가 변하게 된다.

````python
greeting = '안녕하세요?'
greeting = '안녕!'
print(greeting)
````

이번에는 `print(greeting)`이 속한 줄의 순서를 조금 바꾼 다음 실행해보자

````python
greeting = '안녕하세요?'
print(greeting)
greeting = '안녕!'
````

코드는 한 줄씩 순서대로 실행되기 때문에, `print()` 함수가 불리는 시점의 `greeting` 변수의 값인 `안녕하세요?`가 출력된다.

## 4. 함수와 변수 (3) 새로운 함수를 정의하기
이미 만들어진 함수를 부르는 것 외에 우리가 직접 함수를 만들 수도 있다.

부르면 __안녕!__이라고 표시하는 함수를 정의해보겠다. 함수는 내부에서 다른 함수를 부르거나 변수를 정의하는 등 다양한 작업을 할 수 있다. 따라서 함수를 정의하는 방법이 이전의 변수와는 조금 다르다.

참고로, 아래 코드는 실행을 해도 함수를 정의할 뿐 함수 자체를 실행시키지는 않는다. 즉, [Output]에 아무 것도 뜨지 않는 것이 정상이라는 것이다.

````python
def say_hi():
    print('안녕!')
#- 부르면 '안녕!'이라고 인사해주는 함수를 정의합니다.
````

`def 함수명()` 뒤에 `:`를 붙여주고, 이어서 이 함수가 불릴 때 실행할 작업들을 한 줄씩 나열해주면 된다. 단, 해당 함수에 포함된 작업이라는 것을 나타내기 위해, `def` 줄부터 함수가 끝날 때까지 띄어쓰기 4칸으로 들여쓰기를 해주어야 한다.

요약하면, 위 say_hi 함수는 다음과 같이 풀이된다.

_"입력값을 받지 않는 say_hi 함수는 '안녕!'이라는 문자열을 출력해주는 함수입니다."_

`say_hi()` 함수를 불러서 '안녕!'이 나오는 지 한번 확인해보자.

````python
say_hi()
````

````python
def say_hi_no_indent():
print('안녕!')
#- print() 함수 앞에 띄어쓰기 4칸이 빠져있습니다.
````

실행을 해보면 `IndentationError`가 뜨는 것을 확인하실 수 있을 것이다.

이는 들여쓰기된 코드 블럭, 즉 `indented block`을 기대했는데 왜 빼먹었냐는 불평이다. `def` 줄의 다음 줄부터 함수 정의가 끝날 때까지는 들여쓰기를 유지해야 하며, 이 들여쓰기가 유지되는 구간을 블럭(block)이라고 한다. 즉, 파이썬에서 들여쓰기는 이 줄의 코드가 윗 단계의 코드에 속한다는 것을 표시한다.

`안녕!` 다음 줄에 `반가워.`라고 한 줄 더 표시하고 싶다면, 그 다음 줄에 `print()` 함수를 하나 더 추가해주면 된다.

## 5. 함수와 변수 (4) 스코프: 변수의 유효 범위
함수 내에서도 변수를 정의할 수 있다.

하지만 이것은 함수 안에서만 일어난 일이라는 것을 명심해야 한다. 함수가 끝나면 해당 변수는 사라지고, 함수 밖에서 해당 변수를 들여다볼 수도 없다. 이게 무슨 말인지 한번 차근차근 알아보자.

아래 코드를 실행하면 에러가 발생한다.

````python
def say_goodbye():
    goodbye = '잘 가!'

print(goodbye)
````

하지만 반대로 함수 내에서 함수 밖의 변수를 읽는 것은 가능하다.

````python
name = '하루'
#- name이라는 변수는 문자열 '하루'를 가리킵니다.

def say_name():
    print(name)
#- say_name() 함수의 밖에서 정의된 변수(name), 즉 '하루'를 읽어옵니다.

say_name()

name = '시우'
#- 이제, name이라는 변수는 문자열 '시우'를 가리킵니다.

say_name()
````

더 나아가서 잠시 생각해보자. 함수 안에서 `=` 을 통해 값을 바꾸면 실제로 바뀔까? 그렇지 않다. 파이썬은 함수 안에서 새로운 변수를 정의하는 것으로만 이해하기 때문이다. 이게 무슨 말인지 아래 코드를 통해 살펴보자.

````python
name = '하루'
#- name은 문자열 '하루'를 가리키는 변수입니다.

def change_name():
    name = '시우'
    #- 여기서 name은 해당 함수 내에서만 문자열 '시우'를 가리킵니다.

change_name()
#- 함수를 호출해도 아무 일도 일어나지 않습니다.

print(name)
#- 첫 줄, 즉 함수 바깥에서 정의된 문자열 '하루'가 출력됩니다.
````

이처럼 특정 위치에서 어떤 변수에 접근할 수 있는지, 또 한 곳에서 정의된 변수가 어디까지 유효한지 정의된 범위를 변수의 __스코프(scope)__ 라고 한다.

즉, 위의 코드에서 `name = '하루'`는 함수 내부를 포함해 코드 어디에서든 참조할 수 있으므로, __전역 스코프(global scope)__ 를 갖는 변수라고 할 수 있다. 반대로 `name = '시우'`는 `change_name()` 내부에서 정의되어 밖에서 볼 수 없으므로, __지역 스코프(local scope)__ 를 갖는 변수이다.

## 6. 함수와 변수 (5) 함수에 인자 전달하기
만약 원하는 이름을 주고 `[이름], 안녕!`을 표시하게 하려면 어떻게 해야 할까? 아래 코드와 같이 명시적으로 함수에 입력값을 전달해주는 게 바람직하다.

`say_hi_somebody()`라는 새로운 함수를 한번 정의해보자.

````python
def say_hi_somebody(name):
    print(name + ', 안녕!')
#- 함수를 정의합니다. 
#- 이 함수는 문자열인 한 개의 입력값을 받아, '[name], 안녕!'을 출력합니다.
````

그리고 한번 실행해보자.

````python
say_hi_somebody('준이')
#- 위에서 정의한 함수를 호출합니다.
#- "준이, 안녕!"이라고 부르기 위해 입력값으로 '준이'를 주었습니다.
````

그렇다면, `say_hi_somebody(name)` 함수를 찬찬히 살펴보겠다.

함수를 정의하는 라인을 주목해 보면, 괄호 `()` 안에 name이라는 입력값이 정의된 것을 확인할 수 있다. 이 입력값의 자리에 들어오는 값은 해당 함수 내에서 `name`이라는 변수로 불러서 사용할 수 있다. 이처럼 입력값으로 주어진 인자(argument)를 받는 변수를 __매개변수(parameter)__ 라고 한다. 즉, 위 예제에서 함수 내부의 `name`은 매개변수이고, 외부에서 함수를 부를 때 넣어준 `'준이'`는 그 매개변수에 들어가는 인자이다.

만약 위의 함수에 이름을 안 넣고 함수를 부르면 어떻게 될까?

````python
say_hi_somebody()
#- 매개변수를 입력하지 않고 함수를 호출합니다.
````

오류가 난다. `say_hi_somebody()`라는 함수는 `name`이라는 매개변수에 들어갈 인자 없이는 살 수 없는 몸이기 때문이다.

오류 메세지는 `say_hi_somebody()`라는 함수가 `name`이라는 매개변수에 들어갈 필수 위치기반 인자 하나를 누락했다고 말하고 있다. 즉, 우리가 위에서 `say_hi_somebody()` 함수를 정의할 때 `name`이라는 매개변수를 함께 정의했으니, 괄호 속에 들어갈 인자를 빼놓고 부르면 안 된다는 말이다.

이번에는 반대로, 지난 스텝에서 입력값 없이 정의했던 `say_hi_nice()` 함수에 억지로 입력값을 넣어보자.

````python
say_hi_nice('준이')
#- [5-4]에서 정의한 함수를 호출합니다.
````

이번에도 오류가 나지만 내용은 살짝 다르다. `say_hi_nice()` 함수는 위치기반 인자를 받지 않는데 한 개(준이)가 주어졌다는 말이다. 함수를 호출할 때는 미리 정의된 입력값을 넣어야 한다는 교훈을 얻을 수 있다.

그럼 `print()` 함수처럼 입력값이 있던 없던 불평 없이 자기 할 일을 하는 함수를 정의할 수는 없을까? 

````python
def say_hi_default(name='somebody'):
    print('안녕, ' + name + '!')
````

`def` 줄의 `name=` 뒤에 뭔가 붙었다. `say_hi_default()` 함수에 준이를 넣어서 실행해보자.

````python
say_hi_default('준이')
````

이번에는 이름을 뺐을 때 어떻게 동작하는 지 확인해보자

````python
say_hi_default()
````

입력값이 주어지면 그 입력값을 그대로 표시하고, 없으면 `somebody`라는 값이 있는 것처럼 동작한다. 즉, 함수가 받을 매개변수의 기본값을 지정해주면, 입력값이 없을 경우 해당 기본값을 사용하게 된다.

## 7. 함수와 변수 (6) 함수에 인자 여러 개 한꺼번에 전달하기

````python
def say_hi_couple(name1, name2):
    print(name1 + ', ' + name2 + ' 안녕!')
````

인자를 하나 추가하려면 위의 `(name1, name2)`와 같이 괄호 안에서 쉼표로 이름을 다르게 구성해서 넣고, 함수 내에서는 그대로 불러 쓰기만 하면 된다. 확인해보자.

````python
say_hi_couple('그리', '단테')
````

````python
say_hi_couple('그리')
````

## 8. 함수와 변수 (7) 함수의 결과로 값을 반환하기
위에서 함수는 입력값(input)과 출력값(output)을 가질 수 있다고 했다. 위에서 입력값을 가진 함수를 정의하고 사용하는 방법을 살펴보았지만, 사실 여태까지 값을 출력하는 함수는 아직 살펴보지 않았다. 화면에 표시(print)하기는 했지만 말이다.

이게 도대체 무슨 충격적인 소리인지 알아보기 위해서, 일단 여태까지의 방법으로 숫자 두 개를 더하는 함수를 정의해보자.

````python
def add(number1, number2):
    print(number1 + number2)
````

그리고 호출해보자. 일단 잘 동작할 것 같다.

````python
add(1, 2)
````

근데 만약, 이 결과에 숫자를 한 번 더 더하고 싶다면? `print()` 함수 안에서 `add()` 함수를 불러보자.

````python
print(add(1, 2) + 3)
````

일단 위의 코드가 어떤 순서로 실행될지 생각해보자. 마치 수학식처럼, 가장 안쪽의 괄호를 가진 `add(1, 2)`가 먼저 실행되고, 그 이후 옆의 `+ 3`을 더한 뒤, 가장 바깥쪽의 괄호인 `print()`가 실행된다. `add()`로 인해 입력값인 `1`과 `2`의 합인 `3`이 나오고, 거기에 `3`을 다시 더해서 `print()`를 했으니 `6`이 표시되어야 할 것 같다.

하지만 실제 결과는?

세상은 그렇게 호락호락하지 않다. 마음을 추스리고 오류 메세지를 읽어보자. 뭔지는 아직 잘 모르겠지만 `NoneType`과 `int`(정수, integer) 사이에 `+` 연산은 금지라고 한다. 순서상 우리가 `+` 뒤에 쓴 `3`이 정수(`int`)니까, 눈치를 보면 앞의 `NoneType`은 `add(1, 2)`를 가리키는 것일것이다. `Type`은 뭔지 아직 잘 모르겠지만 그 앞의 `None`을 보면 무언가가 없다는 것 같다. 그렇다. `add()`에서 우리는 화면에 표시하라는 `print()` 함수를 부르긴 했지만, 이 함수의 결과가 무엇인지 명시하지 않아서 그렇다.

이처럼 함수가 실행된 뒤 그 결과를 가지고 무언가를 하고 싶다면, `return`으로 이 값을 출력값(output)으로서 반환(return)하라고 명시적으로 표시해줘야 한다.

````python
def add_and_return(number1, number2):
    print(number1 + number2)
    return number1 + number2
````

다시 한번 실행해보자.

````python
print(add_and_return(1, 2) + 3)
````

`print()` 안의 `add_and_return(1, 2)`가 먼저 실행되면서, `add_and_return()` 함수 안 첫 번째 줄인 `print(number1 + number2)`가 실행되어 화면에 `3`이 표시되고, 이후에 `return number1 + number2`로 `3`이 `add_and_return()` 함수의 출력값으로서 반환된다. 그럼 이제 `print(add_and_return(1, 2) + 3)`은 `print(3 + 3)`과 같게 되고, 최종적으로 `print(6)`이 되어 화면에 `6`이 출력된다.

조금 복잡할 수 있으니, 다시 한 번 수학식을 풀 듯 순서대로 풀어보자.

````python
print(add_and_return(1, 2) + 3)
````

계산식을 푸는 것처럼, 우리가 아는 것을 하나씩 순서대로 대입하여, 그 결과값으로 치환하면 된다. 우선 위의 내용 중 괄호의 가장 안쪽인 `add_and_return(1, 2)`가 먼저 수행되어야 한다. 이를 위해 `add_and_return()`이라는 함수의 정의를 보겠다.

````python
def add_and_return(number1, number2):
    print(number1 + number2)
    return number1 + number2
````

## 10. 제어문 (1) `if`: 진실 혹은 거짓
어떤 명제가 참인지 거짓인지에 따라 코드의 실행 여부를 결정할 수 있다.

입력값으로 주어진 숫자가 0 이상일 경우에만 화면에 숫자를 표시하는 함수를 한번 정의해보자.

````python
def print_if_positive(number):
    if number >= 0:
        print(number)
#- 숫자가 0보다 크거나 같을(>=) 경우 즉, 0 이상일 경우에만 숫자를 출력합니다.

print_if_positive(1)
print_if_positive(-1)
````

`def`로 함수를 정의(define)하던 것과 비슷하지만, 이번에는 만약(if)이라는 키워드를 사용한다.

`if 명제:`의 형식으로 표시하며, 이후에 들여쓰기 된 코드 블럭의 내용은 if 키워드 다음의 명제가 참일 때에만 실행된다. 명제를 구성할 때 사용할 수 있는 __비교 연산자(conditional operator)__ 에는 초과 또는 미만(`>`, `<`), 이상 또는 이하(`>=`, `<=`), 값이 같음 또는 다름(`==`, `!=`) 등이 있다.

위의 함수를 변형해서 양이면 `+`, 음이면 `-`를 표시하려면 어떻게 할까?

`if number >= 0:`에 속하는 블럭 뒤에 `if`와 동일한 들여쓰기 깊이로 다시 `if number < 0:`를 추가하는 방법도 있겠지만, 그 이외의 경우(else)를 나타내는 `else:`를 사용해서 앞의 명제가 거짓인 경우에 실행할 코드를 손쉽게 명시할 수 있다.



````python
def print_whether_positive_or_negative(number):
    if number >= 0:
        print('+')
    else:
        print('-')
#- 숫자가 0보다 크거나 같으면 +를, 작으면 -를 출력합니다.

print_whether_positive_or_negative(1)
print_whether_positive_or_negative(-1)
````

그런데 0이 양수(positive)는 아니다.

0에 대한 조건문을 추가하려면 어떻게 할까?

````python
def print_whether_positive_or_negative_or_zero(number):
    if number > 0:
        print('+')
    elif number == 0:
        print('0')
    else:
        print('-')
#- 숫자가 양수이면 +를, 음수이면 -를, 양수도 음수도 아닌 0일 경우 그대로 0을 출력합니다.

print_whether_positive_or_negative_or_zero(1)
print_whether_positive_or_negative_or_zero(0)
print_whether_positive_or_negative_or_zero(-1)
````

위와 같이 else if를 줄인 `elif`로 `if` 뒤, `else` 전에 새로운 조건문을 추가할 수 있다.

난이도를 한 단계 높여보자. 이번엔 양수이면서 짝수일 때만 숫자를 화면에 표시해 보려 한다.

````python
def print_if_positive_and_even(number):
    if (number > 0) and (number % 2 == 0):
        print(number)

#- 숫자가 양수이고(and) 짝수일 때 해당 숫자를 출력합니다.
#-- 양수 : number > 0
#-- 짝수 : number % 2 == 0 즉, 숫자를 2로 나눈 나머지가 0인 경우

print_if_positive_and_even(1)
print_if_positive_and_even(-1)
print_if_positive_and_even(2)
print_if_positive_and_even(-2)
````

위의 코드 중 `%` 연산자는 앞의 숫자를 뒤의 숫자로 나눈 나머지를 의미한다.

보면 두 개의 명제(양수인가 짝수인가)를 `and`(그리고)라는 __논리 연산자(logital operator)__ 로 묶었다. 그 결과로 양쪽 명제가 모두 참일 때만 `if` 조건문에 속한 코드 블럭이 실행된다.

반대로 음수이거나, 혹은 음양에 상관없이 홀수인 경우에만 표시하려면 어떻게 할까?

````python
def print_if_negative_or_odd(number):
    if (number < 0) or (number % 2 != 0):
        print(number)
#- 숫자가 음수이거나(or) 홀수일 때 해당 숫자를 출력합니다.
#- 음수 : number < 0
#- 홀수 : number % 2 != 0 즉, 짝수가 아닐 때

print_if_negative_or_odd(1)
print_if_negative_or_odd(-1)
print_if_negative_or_odd(2)
print_if_negative_or_odd(-2)
````

`or`(또는)이라는 논리 연산자는 양쪽의 명제 중 한 쪽이라도 참일 경우에 작동한다.

## 11. 제어문 (2) `while`: 조건이 참인 동안 반복

````python
def print_if_negative_or_odd(number):
    if (number <= 0) or (number % 2 != 0):
        print(number)
#- [5-8] print_if_negative_or_odd 함수

print_if_negative_or_odd(1)
print_if_negative_or_odd(-1)
print_if_negative_or_odd(2)
print_if_negative_or_odd(-2)
#- print_if_negative_or_odd 함수를 총 4번 호출합니다.
#- 매개변수(number)의 인자를 각각 다르게(1, -1, 2, -2) 부여한 것을 확인할 수 있습니다.
````

위 코드에서 우리는 1, -1, 2, -2를 차례대로 하나하나 입력했었다. 그런데 `while` 반복문을 사용하면 컴퓨터가 자동으로 하나씩 계산하도록 할 수 있다. 참고로, 아래 코드의 `fibonacci()` 함수는 n번째 피보나치 수를 반환하는 함수이다.

````python
def fibonacci(n):
    if n <= 2:
        return 1
    else:
        return fibonacci(n-2) + fibonacci(n-1)

n = 1
while n <= 20:
    print(fibonacci(n))
    n = n + 1
print('끝!')
````

`while` 문은 코드 블럭을 실행하기 전 가장 먼저 조건문을 검사해서, 조건문이 참이면 코드 블럭을 실행한다.

즉, 처음 실행 시 마지막 줄(4행)에서 `n`은 `2`가 되고, `n <= 20`이 참이니 다시 3행으로 돌아가 `print(fibonacci(2))`를 실행한다. 그리고 다시 그 다음 줄에서 `n`은 `3`이 되고, 다시 3행으로 돌아가고.. `n`이 `21`이 되면 그제서야 `while`에 속하는 들여쓰기 코드 블럭을 빠져나와 `print('끝!')`을 실행한다.

이러한 반복문들을 루프(loop, 고리)라고 부른다. 그리고 반복문이 끝없이 계속 실행되는 경우를 무한 루프(infinite loop)라고 부른다. 예를 들어, 아래의 명령을 실행한다면 강제로 종료하기까지 영원히 계속 실행된다.

````python
# 경고 : 실행하지 마세요!
# 만약 실행했다면 터미널 창에서 ctrl+c를 눌러 종료하세요.

while 1 == 1:
    print('Infinite loop!')
````

## 12. 제어문 (3) `for`: 하나씩 반복
파이썬에는 `while` 말고도 `for`이라는 반복문이 있다.

`while` 뒤에는 참과 거짓이 갈릴 수 있는 명제가 온다면, `for` 뒤에는 `a in A`의 형태로, 값이 여러 개 들어있는 목록 `A`에서 값을 하나씩 꺼내서 `a` 변수에 담아 반복문을 실행한다.

설명이 어려우니 한번 예를 들어 이해해보자.

`'Hi'`라는 문자열을 뜯어보면 사실 `'H'`라는 문자와 `'i'`라는 문자가 순서대로 이어지는 형태를 가지고 있다. 즉, `'Hi'` 안에는 `'H'`와 `'i'`라는 값이 하나씩 순서대로 들어있다.



````python
print('H' + 'i' == 'Hi')
````

## 16. 자료형 (3) `NoneType`

````python
print(type(None))
````

앞의 함정들과 다르게 `NoneType` 자료형에 속하는 값은 유한하며, 그마저도 딱 하나이다. 아무것도 없음을 뜻하는 `None`이 바로 그 값이다. `0`도 아무것도 없다는 뜻 아니야?라고 생각되면 아래 도표를 참조.

불교의 공(空) 개념처럼 심오하게 이해할 필요 없이, 그냥 아무것도 없음을 표시하기 위해 만들어놓은 값이다. 이전 예시들 중 함수의 반환값으로 아무것도 정의하지 않았을 경우 `NoneType`과 관련된 오류가 나왔던 것이 바로 이런 상황이다. 이런 상황에서는 `0`과 같은 숫자도, 그 무엇도 반환하지 않았다는 것을 알려주기 위해 `None`이 반환된 것으로 표현한다. 참고로 다른 프로그래밍 언어에서는 같은 개념을 `null`이라고 부르기도 한다.

왜 이런 게 필요할까?<br>
그 어떤 것으로도 정의되지 않는 상태라는 개념은 프로그래밍 시점에 필요한 것이 아니다. 디자인 타임(Design Time)이라고 부르는 설계 타이밍에는 이런 개념이 필요하지 않는다. 모든 개념을 다 정의할 수 있는 시점이기 때문이다.<br>
하지만, 프로그램이 수행하는 도중에 입력값에 따라 전혀 의도하지 않았던 유형의 출력이 발생한다면? 런타임(Runtime)이라고 부르는 프로그램 실행 타이밍에 정의되지 않은 어떤 것이 발생했다고 할때, 그 어떤 것으로도 정의되지 않는 유형을 정의하기 위해 `NoneType`이라는 것이 필요하게 된다.

예를 들어 볼까?

```python
def get_mobile_phonenum(value):
    if len(value) >= 11:     # 010XXXXYYYY 의 길이는 11 이상
        return new PhoneNumber(value)
```

어떤 스트링 value로부터 모바일 전화번호 유형의 객체를 구하는 메소드를 위와 같이 짰다고 해보자. 그런데 저 __코드를 돌리는 도중에__ value에 10자리 숫자가 입력되었다고 한다. 그 순간 에러를 내면서 프로그램이 죽어야 할까? 아니면 NoneType 객체라도 리턴하는게 좋을까?

이것은 인터프리터 언어인 파이썬의 설계사상과 관련이 있다. 컴파일 타임에 모든 유형이 다 정의되어야 하는 C같은 언어는 NoneType을 필요로 하지 않는다. 정의될 수 없는 어떤 것이 나온다면 컴파일 오류를 내면 되기 때문이다. 하지만 어떤 것의 유형이 런타임에 정의되어도 무방하다거나, 심지어 __늦게 정의될 수록 좋다고 생각하는 파이썬 같은 언어__는 `NoneType` 유형을 필요로 하게 된다.

## 17. 자료형 (4) 불리언_boolean, `bool`

## 19. 자료형 (6) 튜플과 리스트_`tuple`, `list`

튜플(tuple)과 리스트(list)이다.

__튜플__의 경우에는 괄호`()` 안에 원하는 값들을 쉼표(`,`)로 구분해서 써주면 된다.

````python
numbers = (1, 2, 3)
for number in numbers:
    print(number)
````

물론 숫자만 들어가라는 법은 없고, 문자든 어떤 값이든 넣어도 됩니다.

````python
mixed_tuple = ('Hello', 0, True)
````

튜플도 문자열처럼 `[색인]` 문법을 써서 특정 위치의 값을 읽을 수 있다.

````python
tuple_123 = (1, 2, 3)
print(tuple_123[0])
````

하지만 값을 변경할 수는 없다. 아래 코드는 에러가 난다.

````python
tuple_123 = (1, 2, 3)
tuple_123[0] = 0
````

튜플 안에 튜플을 넣을 수도 있다.

이런 경우에는 `[n]` 문법을 연이어 써서 튜플 안의 튜플 안의 값을 가져올 수 있다.

````python
nested_tuple = ((1, 2), (3, 4))
print(nested_tuple[1][0])
````

튜플끼리 덧셈 연산자를 통해 합칠 수도 있다.

````python
tuple_a = (1, 2)
tuple_b = (3, 4)
print(tuple_a + tuple_b)
````

`in` 연산자를 써서 특정한 값이 튜플 안에 있는지 없는지도 검사할 수 있다.

````python
tuple_abc = ('a', 'b', 'c')
print('a' in tuple_abc)
print('d' in tuple_abc)
````

__리스트__는 튜플과 거의 같다. 괄호만 대괄호(`[`, `]`)로 바꿔주면 된다.

````python
list_a = [1, 2]
list_b = ['3', '4']
print(list_a, list_b)
````

그리고 튜플과 달리 값을 바꿀 수도 있다.

````python
list_123 = [1, 2, 3]
list_123[0] = 0
print(list_123)
````

리스트만이 가진 또 다른 기능이 있다.

변수를 한번 정의해놓고 변수명 뒤에 `.append()`라는 특별한 함수를 뒤에 붙여서 값을 하나씩 추가하거나, `.remove()`를 통해 특정 값을 빼거나, `.pop()`을 통해 특정 순서의 값을 뺄 수 있다는 점이다.

## 20. 자료형 (7) 딕셔너리_dictionary, `dict`
딕셔너리(dictionary)도 값을 여러 개 포함하는 컨테이너 자료형의 일종이다.

리스트와 차이점은 숫자로 된 인덱스를 쓰는 대신, 직접 인덱스를 원하는 대로 지정할 수 있다는 점이다.

````python
conductor = {'first_name': '단테', 'last_name': '안'}
print(conductor['first_name'])
````

위의 코드에서 `'first_name'`와 `'last_name'`은 각각 인덱스 또는 딕셔너리의 키(key)들이고, `'단테'`와 `'안'`은 각각 해당 키의 값(value)이다. 문법적으로는 중괄호(`{`, `}`) 안에서 쉼표(`,`)로 각 항목(item)을 나누고, 항목 별로 쌍점(`:`)을 써서 키와 값을 나눈다.

딕셔너리에 새로운 값을 추가하고 싶다면 그냥 새로운 키를 넣으면 된다.

````python
conductor = {'first_name': '단테', 'last_name': '안'}
conductor['gender'] = 'male'
print(conductor)
````

제거하고 싶다면 리스트와 마찬가지로 `.pop()`을 사용하면 된다.

````python
conductor = {'first_name': '단테', 'last_name': '안'}
conductor.pop('last_name')
print(conductor)
````

딕셔너리에 `for` 루프문을 사용하면 기본적으로 키 값만 돌려준다.

````python
conductor = {'first_name': '단테', 'last_name': '안'}
for key in conductor:
    print(key)
````

하지만 `.items()`를 사용해서 키와 값을 둘 다 뽑아올 수도 있다.

````python
conductor = {'first_name': '단테', 'last_name': '안'}
for key, value in conductor.items():
    print(key + ' : ' + value)
````

## 24. 더 빠르게! (심화)
만들어진 피보나치 함수에 한가지 문제가 있다. 최초의 `n`이 커질수록 속도는 기하급수적으로 느려지는 점이다.

우선 왜 느려지는지 한번 생각해보자. 우선 입력값이 `1`, `2`일 경우에는 `if`문에 의해 `fibonacci()` 안에서 다시 `fibonacci()`를 부르는 일 없이 실행이 끝난다. 하지만 입력값이 `3`이면, `else`문으로 빠져 `fibonacci()` 함수를 2번 더 불러야 한다. 입력값이 `4`이면, `fibonacci(3)`과 `fibonacci(2)`를 불러야 하고, 여기서 `fibonacci(3)` 때문에 다시 `fibonacci(2)`와 `fibonacci(1)`을 또 부르고… 결국 우리가 짠 코드 상으로는 입력값 `n`에 대해 `fibonacci()`함수를 대략 2의 `n` 제곱 번 불러야 한다.

계산 횟수 자체도 문제지만, 한 번 불린 함수/입력값 조합이 반복적으로 불리기 때문에 이미 계산한 값을 다시 계산해야 한다는 비효율성도 있다. 한 번 계산된 값은 또 계산하지 않도록, 예를 들어 `fibonacci(40)`을 부를 때 `fibonacci(4)`의 값을 기억해 두고 처음 한 번만 계산하는 방법은 없을까?

변수의 스코프를 활용하면 가능하다. 함수 밖에서 딕셔너리를 만들어서 한 번 계산된 값들을 기억하도록 만들어보자. 딕셔너리의 이름은 `memory`로 하고, 피보나치 순서 `n`이 키, 그 순서에 해당하는 피보나치 숫자가 값이 되도록 한다. 여기에 일단 우리가 알고 있는 1번째 피보나치 수 1, 2번째 피보나치 수 1을 알려준다.

````python
memory = {1: 1, 2: 1}
````

이건 함수 밖에서 정의한 변수이니 함수 안에서 가져다 쓸 수 있고, 함수가 실행이 종료되어도 계속 기억된다. 이러한 성질을 이용해서, 이미 기억에 있는 숫자 `n`이 들어올 경우 계산하지 말고 기억에 있는 그대로 읽어서 내보내라고 한다.

````python
memory = {1: 1, 2: 1}

def fibonacci(n):
    if n in memory:
        number = memory[n]
    else:
        number = fibonacci(n-1) + fibonacci(n-2)
    return number
````

`if`문에 딸린 조건문과 코드 블럭이 변경되었다. `if n in memory`를 통해 입력값 `n`이라는 키가 `memory` 딕셔너리에 이미 있는 경우, 그 값 `memory[n]`을 바로 `number`로써 반환하라는 이야기이다.

기억을 읽어올 준비는 되었으니, 한 번 계산한 숫자를 기억에 쓰는 부분을 작성하자.

````python
memory = {1: 1, 2: 1}

def fibonacci(n):
    if n in memory:
        number = memory[n]
    else:
        number = fibonacci(n-1) + fibonacci(n-2)
        memory[n] = number
    return number
````

이제 `fibonacci(n-1) + fibonacci(n-2)`를 통해 구한 값은 `memory`에 저장된다. `memory` 변수를 새로 정의하는 것이 아니라 이미 있는 딕셔너리에 새로운 값을 추가하는 것이니, 함수 내부에서 똑같은 이름의 변수가 새로 만들어지는 문제도 피할 수 있다. 이제 한번 큰 숫자를 돌려서 체감상 차이가 있는지 한번 보자. 그리고 실제로 `memory`에 잘 저장이 되었는지 찍어보자.

````python
memory = {1: 1, 2: 1}

def fibonacci(n):
    if n in memory:
        number = memory[n]
    else:
        number = fibonacci(n-1) + fibonacci(n-2)
        memory[n] = number
    return number

print(fibonacci(100))

print(memory)
````

부드럽게 돌아간다. 프로그래밍에서 이렇게 중간 계산값을 기억해놓고, 다시 계산하는 대신 값을 바로 읽어 쓰는 방식으로 계산 시간을 줄이는 기법을 __메모이제이션(memoization)__ 이라고 한다.
